### Aufgabe 5

Gegeben sind die gleichen Daten, wie in Aufgabe 4. 

a)	Stelle diese Daten in einem geeigneten Diagramm dar. Begründe deine Wahl.

b)	Ergänze deine Grafik mit der Regressionsgeraden.

c)	Erkennst du aus dieser Grafik einen Zusammenhang zwischen den beiden Merkmalen? Wenn ja, wie ist das zu interpretieren?

d)	Lies aus der Grafik (Interpolation) den ungefähren Wert für die Temperatur nach sieben Betriebsstunden ab.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress
# Gegebene Daten
betriebsstunden = np.array([1.5, 3.6, 4.1, 4.8, 5.6, 6.3, 9.2, 10.1])
temperaturen = np.array([32.1, 35.4, 36.2, 36.5, 37.2, 39.7, 41.3, 43.5])
# a) Daten in einem geeigneten Diagramm darstellen
# Ein Streudiagramm ist geeignet, da es die Beziehung zwischen zwei kontinuierlichen Variablen zeigt.
plt.scatter(betriebsstunden, temperaturen, color='blue', label='Datenpunkte')
# b) Regressionsgerade ergänzen
slope, intercept, r_value, p_value, std_err = linregress(betriebsstunden, temperaturen)
regressionsgerade = slope * betriebsstunden + intercept
plt.plot(betriebsstunden, regressionsgerade, color='red', label='Regressionsgerade')
# c) Zusammenhang interpretieren
korrelationskoeffizient = r_value
if korrelationskoeffizient > 0.7:
    interpretation = "starker positiver Zusammenhang"
elif korrelationskoeffizient > 0.5:
    interpretation = "moderater positiver Zusammenhang"
elif korrelationskoeffizient > 0.3:
    interpretation = "schwacher positiver Zusammenhang"
else:
    interpretation = "kein oder sehr schwacher Zusammenhang"
print(f"Korrelationskoeffizient (r): {korrelationskoeffizient}")
print(f"Interpretation: {interpretation}")
# d) Interpolation für 7 Betriebsstunden
betriebsstunden_interpol = 7
temperatur_interpol = slope * betriebsstunden_interpol + intercept
print(f"Interpolierte Temperatur bei 7 Betriebsstunden: {temperatur_interpol:.2f} ℃")
# Ergänzungen zur Grafik
plt.scatter(betriebsstunden_interpol, temperatur_interpol, color='green', label=f'Interpolierter Wert bei 7h: {temperatur_interpol:.2f} ℃')
plt.xlabel('Betriebsstunden [h]')
plt.ylabel('Temperatur [℃]')
plt.title('Zusammenhang zwischen Betriebsstunden und Temperatur')
plt.legend()
plt.show()
```

Kommentar zu Aufgabe a)  Ein Streudiagramm (`plt.scatter`) ist geeignet, da es die Beziehung zwischen zwei kontinuierlichen Variablen(Betriebsstunden und Temperatur) zeigt.

Kommentar zu Aufgabe c)  korrelationskoeffizient > 0.7 Interpretation: starker positiver Zusammenhang.

Kommentar zu Aufgabe d) 39.41°C
