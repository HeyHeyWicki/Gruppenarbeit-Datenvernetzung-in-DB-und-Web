a)  Die in Aufgabe 2 generierten Daten sollen mit Hilfe der Methode ¬´pd.read_csv()¬ª aus dem Modul/Bibliothek ¬´pandas¬ª, eingelesen werden. 

b)  F√ºge nun den Daten eine Spalten√ºberschrift ¬´ZufallsZahl¬ª hinzu und gib die ersten sechs Zeilen der Daten aus.

c)  Da die Daten mit Hilfe der ¬´pandas¬ª-Methode ¬´read_csv()¬ª eingelesen worden sind, gibt es nun eine M√∂glichkeit, sich schnell einen √úberblick √ºber die Daten zu beschaffen. Der Befehl heisst ¬´describe¬ª. Lass dir dadurch die wichtigsten Parameter anzeigen. Welche Bedeutung haben die einzelnen Parameter?

d)  Erstelle mit Hilfe der Pandas-Methode ein Histogramm. Die Anzahl der Klassen ùëò wird in der Regel durch die Formel ùëò = ‚åä‚àöùëõ‚åã festgelegt.




```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
# a) Daten einlesen
filename = 'daten_p3_2.txt'
data = pd.read_csv(filename, header=None)
# b) Spalten√ºberschrift hinzuf√ºgen und die ersten sechs Zeilen ausgeben
data.columns = ['ZufallsZahl']
print(data.head(6))
# c) √úberblick √ºber die Daten mit describe()
description = data.describe()
print(description)
# Bedeutung der Parameter:
# - count: Anzahl der Datenpunkte
# - mean: Mittelwert der Daten
# - std: Standardabweichung der Daten
# - min: Minimalwert der Daten
# - 25%: 25. Perzentil (unteres Quartil)
# - 50%: Median (50. Perzentil)
# - 75%: 75. Perzentil (oberes Quartil)
# - max: Maximalwert der Daten
# d) Histogramm erstellen
n = len(data)
k = int(np.sqrt(n))
data['ZufallsZahl'].plot(kind='hist', bins=k, edgecolor='black')
plt.xlabel('Wert')
plt.ylabel('H√§ufigkeit')
plt.title('Histogramm der normalverteilten Zufallszahlen')
plt.show()
```

Kommentar zu Aufgabe C) Die Bedeutung der Parameter:
     - `count`: Anzahl der Datenpunkte.
     - `mean`: Mittelwert der Daten.
     - `std`: Standardabweichung der Daten.
     - `min`: Minimalwert der Daten.
     - `25%`: 25. Perzentil (unteres Quartil).
     - `50%`: Median (50. Perzentil).
     - `75%`: 75. Perzentil (oberes Quartil).
     - `max`: Maximalwert der Daten.
