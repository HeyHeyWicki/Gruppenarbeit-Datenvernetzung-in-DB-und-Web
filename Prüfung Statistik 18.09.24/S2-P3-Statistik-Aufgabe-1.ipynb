### Aufgabe 1

Eine Münze soll n-mal geworfen werden. Dabei soll jedes Ereignis «Kopf» oder «Zahl» gleichwahrscheinlich auftreten.

a)  Erstelle ein Pythonprogramm, welches dir diese Daten generiert und in die Datei «daten_p3_1_out.csv» speichert.

b)  Analysiere diese Daten, zähle, wie oft die beiden Ereignisse «Kopf» und «Zahl» aufgetreten sind.

c)  Erstelle ein Histogramm, welches diese Häufigkeiten visualisiert.

d)  Vergleiche diese Häufigkeiten mit der erwarteten (theoretischen Wahrscheinlichkeit) absolut und relativ. Gibt es Unterschiede – warum?


```python
import random
import csv
import matplotlib.pyplot as plt
# Anzahl der Würfe
n = 1000
# a) Daten generieren und in Datei speichern
def generate_coin_tosses(n):
    results = []
    for _ in range(n):
        toss = 'Kopf' if random.random() < 0.5 else 'Zahl'
        results.append(toss)
    return results
def save_to_csv(data, filename):
    with open(filename, mode='w', newline='') as file:
        writer = csv.writer(file)
        writer.writerow(['Wurf', 'Ergebnis'])
        for i, result in enumerate(data):
            writer.writerow([i + 1, result])
# b) Daten analysieren
def analyze_data(data):
    counts = {'Kopf': 0, 'Zahl': 0}
    for result in data:
        counts[result] += 1
    return counts
# c) Histogramm erstellen
def plot_histogram(counts):
    labels = list(counts.keys())
    values = list(counts.values())
    plt.bar(labels, values, color=['blue', 'orange'])
    plt.xlabel('Ergebnis')
    plt.ylabel('Häufigkeit')
    plt.title('Häufigkeit von Kopf und Zahl')
    plt.show()
# d) Häufigkeiten vergleichen
def compare_frequencies(counts, n):
    expected_frequency = n / 2
    print(f"Erwartete Häufigkeit für Kopf: {expected_frequency}")
    print(f"Erwartete Häufigkeit für Zahl: {expected_frequency}")
    print(f"Absolute Häufigkeit für Kopf: {counts['Kopf']}")
    print(f"Absolute Häufigkeit für Zahl: {counts['Zahl']}")
    print(f"Relative Häufigkeit für Kopf: {counts['Kopf'] / n}")
    print(f"Relative Häufigkeit für Zahl: {counts['Zahl'] / n}")
# Hauptprogramm
if __name__ == "__main__":
    # Daten generieren
    data = generate_coin_tosses(n)
    # Daten speichern
    save_to_csv(data, 'daten_p3_1_out.csv')
    # Daten analysieren
    counts = analyze_data(data)
    # Histogramm erstellen
    plot_histogram(counts)
    # Häufigkeiten vergleichen
    compare_frequencies(counts, n)
```

Kommentar: Die Absolute Häufigkeit für Zahl ist etwas höher als Kopf. Das heisst Zahl ist genau 509mal vorgekommen. Erwartet ist 50 -50 da beide Kopf oder Zahl gleich viel vorkommen kann.
