### Aufgabe 6

Gegeben sind die Daten, welche in der Datei «daten_p1_6.csv» vorliegen. Die beiden Merkmale sind durch das Trennzeichen «;» voneinander getrennt.

a)	Lies diese Daten ein und lass dir die ersten fünf Datensätze anzeigen, damit du dir ein Bild über die Struktur der Daten und die Merkmale machen kannst.

b)	Berechne die beiden Koeffizienten der Regressions- respektive Ausgleichsgeraden.

c)	Wie gross ist der Zusammenhang (Korrelation) zwischen den beiden Merkmalen? Berechne dazu den Korrelationskoeffizienten und interpretiere das Resultat.


```python
import pandas as pd
import numpy as np
from scipy.stats import linregress
# a) Daten einlesen und die ersten fünf Datensätze anzeigen
filename = 'daten_p1_6.csv'
data = pd.read_csv(filename, delimiter=';')
print(data.head(5))
# Annahme: Die Datei hat zwei Spalten, die wir als 'Merkmal1' und 'Merkmal2' bezeichnen
merkmal1 = data.iloc[:, 0]
merkmal2 = data.iloc[:, 1]
# b) Berechnung der Regressionskoeffizienten
slope, intercept, r_value, p_value, std_err = linregress(merkmal1, merkmal2)
print(f"Steigung (b): {slope}")
print(f"Achsenabschnitt (a): {intercept}")
# c) Berechnung des Korrelationskoeffizienten
korrelationskoeffizient = r_value
print(f"Korrelationskoeffizient (r): {korrelationskoeffizient}")
# Interpretation des Korrelationskoeffizienten
if korrelationskoeffizient > 0.7:
    interpretation = "starker positiver Zusammenhang"
elif korrelationskoeffizient > 0.5:
    interpretation = "moderater positiver Zusammenhang"
elif korrelationskoeffizient > 0.3:
    interpretation = "schwacher positiver Zusammenhang"
else:
    interpretation = "kein oder sehr schwacher Zusammenhang"
print(f"Interpretation: {interpretation}")
```

Kommentar zu Aufgabe c) korrelationskoeffizient < 0.3 kein oder sehr schwacher Zusammenhang.
