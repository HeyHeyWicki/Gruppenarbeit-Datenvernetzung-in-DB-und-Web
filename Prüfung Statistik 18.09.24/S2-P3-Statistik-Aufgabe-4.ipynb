a)	Berechne die beiden Koeffizienten der Regressions- respektive Ausgleichgeraden.

b)	Wie gross ist der Zusammenhang zwischen den Betriebsstunden und den Temperaturen? Berechne dazu den Korrelationskoeffizienten und interpretiere das Resultat.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress
# Gegebene Daten
betriebsstunden = np.array([1.5, 3.6, 4.1, 4.8, 5.6, 6.3, 9.2, 10.1])
temperaturen = np.array([32.1, 35.4, 36.2, 36.5, 37.2, 39.7, 41.3, 43.5])
# a) Berechnung der Regressionskoeffizienten
slope, intercept, r_value, p_value, std_err = linregress(betriebsstunden, temperaturen)
print(f"Steigung (b): {slope}")
print(f"Achsenabschnitt (a): {intercept}")
# Regressionsgerade
regressionsgerade = slope * betriebsstunden + intercept
# b) Berechnung des Korrelationskoeffizienten
korrelationskoeffizient = r_value
print(f"Korrelationskoeffizient (r): {korrelationskoeffizient}")
# Interpretation des Korrelationskoeffizienten
if korrelationskoeffizient > 0.7:
    interpretation = "starker positiver Zusammenhang"
elif korrelationskoeffizient > 0.5:
    interpretation = "moderater positiver Zusammenhang"
elif korrelationskoeffizient > 0.3:
    interpretation = "schwacher positiver Zusammenhang"
else:
    interpretation = "kein oder sehr schwacher Zusammenhang"
print(f"Interpretation: {interpretation}")
# Visualisierung der Daten und der Regressionsgerade
plt.scatter(betriebsstunden, temperaturen, color='blue', label='Datenpunkte')
plt.plot(betriebsstunden, regressionsgerade, color='red', label='Regressionsgerade')
plt.xlabel('Betriebsstunden [h]')
plt.ylabel('Temperatur [℃]')
plt.title('Regressionsanalyse der Betriebsstunden und Temperaturen')
plt.legend()
plt.show()
```
