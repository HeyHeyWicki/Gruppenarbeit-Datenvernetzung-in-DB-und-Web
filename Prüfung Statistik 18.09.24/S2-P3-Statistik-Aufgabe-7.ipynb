### Aufgabe 7

Gegeben sind die gleichen Daten, wie in Aufgabe 6. Diese sind in dieser Datei gespeichert «daten_p1_6.csv» Die beiden Merkmale sind durch das Trennzeichen «;» voneinander getrennt.

a)	Stelle diese Daten in einem geeigneten Diagramm dar. Begründe deine Wahl.

b)	Ergänze deine Grafik mit der Regressionsgeraden.

c)	Erkennst du einen Zusammenhang zwischen den beiden Merkmalen? Wenn ja, wie ist das zu interpretieren?


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress
# Daten einlesen
filename = 'daten_p1_6.csv'
data = pd.read_csv(filename, delimiter=';')
# Annahme: Die Datei hat zwei Spalten, die wir als 'Merkmal1' und 'Merkmal2' bezeichnen
merkmal1 = data.iloc[:, 0]
merkmal2 = data.iloc[:, 1]
# a) Daten in einem geeigneten Diagramm darstellen
# Ein Streudiagramm ist geeignet, da es die Beziehung zwischen zwei kontinuierlichen Variablen zeigt.
plt.scatter(merkmal1, merkmal2, color='blue', label='Datenpunkte')
# b) Regressionsgerade ergänzen
slope, intercept, r_value, p_value, std_err = linregress(merkmal1, merkmal2)
regressionsgerade = slope * merkmal1 + intercept
plt.plot(merkmal1, regressionsgerade, color='red', label='Regressionsgerade')
# c) Zusammenhang interpretieren
korrelationskoeffizient = r_value
if korrelationskoeffizient > 0.7:
    interpretation = "starker positiver Zusammenhang"
elif korrelationskoeffizient > 0.5:
    interpretation = "moderater positiver Zusammenhang"
elif korrelationskoeffizient > 0.3:
    interpretation = "schwacher positiver Zusammenhang"
else:
    interpretation = "kein oder sehr schwacher Zusammenhang"
print(f"Korrelationskoeffizient (r): {korrelationskoeffizient}")
print(f"Interpretation: {interpretation}")
# Ergänzungen zur Grafik
plt.xlabel('Merkmal 1')
plt.ylabel('Merkmal 2')
plt.title('Zusammenhang zwischen Merkmal 1 und Merkmal 2')
plt.legend()
plt.show()
```

Kommentar zu Aufgabe a) Ein Streudiagramm ist geeignet, da es die Beziehung zwischen zwei kontinuierlichen Variablen zeigt.

Kommentar zu Aufgabe c) DA der Korrelationskoeffizient kleiner als < 0.3 ist, besteht kein oder sehr schwacher Zusammenhang.
