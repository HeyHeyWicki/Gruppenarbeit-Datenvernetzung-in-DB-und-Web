## Aufgabe 1
### a) Türstatus
Ein Türsensor gibt den Status einer Tür an ("offen" oder "geschlossen"). Schreibe ein Programm, das den aktuellen Status überprüft und eine Meldung ausgibt. Das Programm soll dich nach einem möglich Status befragen.

````python
# Türstatus abfragen
tuer_status = input('Gib einen Türstatus ein: "offen" / "geschlossen":')

# Selektion zur Überprüfung des Türstatus
if tuer_status == 'offen':
    print('Achtung: Die Tür ist offen!')
else:
    print('Die Tür ist geschlossen.')
````

### b) Türstatus mit Gültigkeitsprüfung
Ein Türsensor gibt den Status einer Tür an ("offen" oder "geschlossen"). Schreibe ein Programm, das den aktuellen Status überprüft und eine Meldung ausgibt. Das Programm soll dich nach einem möglich Status befragen und die Eingabe überprüfen, ob ein gültiger Status eingegeben wurde.

````python
# Gültige Stati festlegen
gueltiger_status = ['offen', 'geschlossen']

gueltig = False
while not(gueltig):
    tuer_status = input('Gib einen Türstatus ein: "offen" / "geschlossen":')
    if tuer_status in gueltiger_status:
        gueltig = True
        
# Selektion zur Überprüfung des Türstatus
if tuer_status == 'offen':
    print('Achtung: Die Tür ist offen!')
else:
    print('Die Tür ist geschlossen.')       
    
````

## Aufgabe 2
### a) Anzahl offene Türen
Eine Liste speichert den Status von mehreren Türen im Gebäude ("offen" oder "geschlossen"). Schreibe ein Programm, das mit einer for-Schleife zählt, wie viele Türen offen sind. Definiere dir eine Variable, in welcher du die verschiedenen Stati eintragen kannst: 
```python 
tuer_status_liste = ['offen', 'geschlossen', 'geschlossen', 'offen', 'offen']
```
Anschliessend soll das Programm diese Stati zählen und ausgeben.


````python
# Status der Türen
tuer_status_liste = ['offen', 'geschlossen', 'geschlossen', 'offen', 'offen']

# Zähler für offene Türen
offene_tueren = 0

# Iteration zur Zählung der offenen Türen
for status in tuer_status_liste:
    if status == 'offen':
        offene_tueren = offene_tueren + 1       # oder: offene_tueren += 1

print('Anzahl der offenen Türen:', offene_tueren)
````

### b) Anzahl offene Türen - eigene Eingabe
Erweitere dein Programm so, dass du selber für mehrere Räume die gültigen Stati eingeben kannst. Der Abbruch soll durch die Eingabe "stop" erfolgen. Die eingegebenen Statis sollen in einer Liste gespeichert werden. 

````python
# Status der Türen
tuer_status_liste = []

abbruch = False
while not(abbruch):
    status = input('Gib den Status der Türe ein (Abbruch: "stop"):')
    if status == 'stop':
        abbruch = True
    else:
        tuer_status_liste.append(status)
            
# Zähler für offene Türen
offene_tueren = 0

# Iteration zur Zählung der offenen Türen
for status in tuer_status_liste:
    if status == 'offen':
        offene_tueren = offene_tueren + 1       # oder: offene_tueren += 1

print('Status:', tuer_status_liste, 'Anzahl der offenen Türen:', offene_tueren)
````

### c) Anzahl offene Türen - eigene Eingabe und Gültigkeitsprüfung
Erweitere dein Programm so, dass die eingegebenen Stati gültig sein müssen. 

````python
# Gültige Stati festlegen
gueltiger_status = ['offen', 'geschlossen']
tuer_status_liste = []

abbruch = False
while not(abbruch):
    gueltig = False
    while not(gueltig):
        tuer_status = input('Gib einen Türstatus ein: "offen" / "geschlossen", Abbruch: "stop":')
        if tuer_status in gueltiger_status:
            tuer_status_liste.append(tuer_status)
        elif tuer_status == 'stop':
            abbruch = True
            gueltig = True
                
# Iteration zur Zählung der offenen Türen
offene_tueren = 0
for status in tuer_status_liste:
    if status == 'offen':
        offene_tueren = offene_tueren + 1             
            
print('Status:', tuer_status_liste, 'Anzahl der offenen Türen:', offene_tueren)            
````

## Aufgabe 3



Erstelle ein Pythonprogramm, welches die Quersumme einer natürlichen Zahl berechnet. Der Benutzer soll nach dieser gefragt werden.

__Hinweis__: Die Zahlen, respektive deren Ziffern, sind in effektive Zahlen umzuwandeln, bevor man mit ihnen rechnen kann!

````python
zahl = input('Gib eine natürliche Zahl ein: ')

str_zahl = zahl
quersumme = 0
for ziffer in str_zahl:
    # jedes Zeichen in eine Zahl umwandeln und addieren
    quersumme = quersumme + int(ziffer)

print('Die Quersumme der Zahl', zahl, 'lautet: ', quersumme)
````

## Aufgabe 5
Selektion, Iteration, Liste, ganzzahlige Division.

An Hand der Ulam-Folge wollen wir die Kontrollstrukturen __Selektion__ und __Repetition__ üben. Siehe dazu auch [Wikipedia](https://de.wikipedia.org/wiki/Ulam-Folge). Dieses Problem ist bis heute unbewiesen. Man vermutet, dass unabhängig von der gewählten Startzahl, der Algorithmus bei 1 endet. Der Algorithmus kann wie folgt beschrieben werden:

> 1.    Wähle als Startzahl n eine beliebige natürliche Zahl.
> 2.    Für gerades n entsteht das nächste Element der Folge, indem man n halbiert.
> 3.    Für ungerades n muss man n entsprechend verdreifachen und dann 1 addieren.
> 4.    Wiederhole die Schritte 2 oder 3.
> 
### a) ohne Liste

````python
start_zahl = int(input('Gib Startzahl ein:'))

zahl = start_zahl
while zahl > 1:
    if zahl % 2 == 0:
        zahl = zahl // 2
    else:
        zahl = 3 * zahl + 1
    print(zahl)

````

### b) mit Liste

````python
zahl = int(input('Gib Startzahl ein:'))

zahlen = [zahl]
while zahl > 1:
    if zahl % 2 == 0:
        zahl = zahl // 2
    else:
        zahl = 3 * zahl + 1
    zahlen.append(zahl)

print('Die Ulam-Folge lautet:', zahlen)
````

## Aufgabe 6
### a) ohne enumerate
Ein Gebäudeautomationssystem überwacht verschiedene Räume, um festzustellen, ob die Lüftung eingeschaltet werden muss. Schreibe ein Programm, das eine Liste von CO2-Werten in ppm für verschiedene Räume durchläuft. Wenn ein Wert über 1000 ppm liegt, soll das Programm eine Warnmeldung für diesen Raum ausgeben, andernfalls soll eine Meldung ausgegeben werden, dass die Luftqualität akzeptabel ist.

````python
# Liste von CO2-Werten für verschiedene Räume in ppm
co2_werte = [950, 1020, 870, 1100, 990]
schwellen_wert = 1000

# Zähler für die Raumindizes
raum_nummer = 1

# Iteration über die CO2-Werte der Räume
for wert in co2_werte:
    if wert > schwellen_wert:
        print('Raum:', raum_nummer, 'Warnung! CO₂-Wert zu hoch:', wert, 'ppm')
    else:
        print('Raum:', raum_nummer, 'Luftqualität ist akzeptabel:', wert, 'ppm')
    raum_nummer = raum_nummer + 1  # Erhöhe den Zähler für den nächsten Raum
````

### b) mit __enumerate__
Erweitere dein Programm, so dass es mit einer Schleife und zwei Schleifen-Variablen auskommt. Verwende dazu die Funktion [enumerate](https://docs.python.org/3/library/functions.html#enumerate).

````python
# Liste von CO2-Werten für verschiedene Räume in ppm
co2_werte = [950, 1020, 870, 1100, 990]
schwellen_wert = 1000

# Iteration über die CO2-Werte der Räume
for index, wert in enumerate(co2_werte):
    print(index, wert)
    if wert > schwellen_wert:
        print(f"Raum {index + 1}: Warnung! CO₂-Wert zu hoch ({wert} ppm)")
    else:
        print(f"Raum {index + 1}: Luftqualität ist akzeptabel ({wert} ppm)")
````

## Aufgabe 8
### a) Fakultät - mit vorhandener Funktion
Oft werden die Anzahl der [Permutationen](https://studyflix.de/statistik/permutation-ohne-wiederholung-1071) ohne Wiederholung benötigt. Wir können dies einfach durch eine vorhandene Funktion aus der Bibliothek __math__ mit der Funktion __factorial__ berechnen. Das Programm soll nach einer natürlichen Zahl abfragen und anschliessend die Fakültät ausgeben.

````python
from math import factorial

# Eingabe
zahl = int(input('Berechne die Fakultät von: '))

# Ausgabe der Fakultät              
print('Die Fakulät von', zahl, 'ist:', factorial(zahl))    
````

### b) Fakultät - iterativ
Jetzt soll das Programm angepasst werden so, dass wir die Berechnung mit Hilfe einer Iteration durchführen. Dazu müssen wir uns überlegen, wie die [Fakultät](https://de.wikipedia.org/wiki/Fakultät_(Mathematik)) definiert ist.

````python
zahl = int(input('Berechne die Fakultät von: '))

# das Produkt wird initialisiert
fak = 1

# wichtig: obere Grenze läuft standargemäss nur bis zahl-1
for index in range(1, zahl + 1):
    # das Produkt wird faktorweise erweitert    
    fak = fak * index

# Ausgabe der Fakultät              
print('Die Fakulät von', zahl, 'ist:', fak)    
````

## Aufgabe 9
Fibonacci-Zahlen - iterativ. Berechne die [Fibonacci-Zahlen](https://de.wikipedia.org/wiki/Fibonacci-Folge). Diese erscheinen oft in der [Natur](https://de.wikipedia.org/wiki/Fibonacci-Folge#Fibonacci-Folgen_in_der_Natur) und [Architektur](https://uol.de/f/5/inst/mathe/download/fibonacci.pdf).


````python
zahl = int(input('Welche ersten Fibonacci-Zahlen sollen ausgegeben werden?'))

a = 0
b = 1
resultat = []
for i in range(zahl):
    resultat = resultat + [a]
    c = a + b
    a = b
    b = c
    
print('Die Fobonacci-Zahlen lauten:', resultat)
````
