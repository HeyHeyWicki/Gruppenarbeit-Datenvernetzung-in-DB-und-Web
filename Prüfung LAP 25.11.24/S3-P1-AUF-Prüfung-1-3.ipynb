### Aufgabe 3: Belueftungssysteme mit Benutzerinteraktion
Ein Belüftungssystem soll die Luftfeuchtigkeit in fünf Räumen überwachen. Der Benutzer gibt die Luftfeuchtigkeitswerte der Räume in Prozent ein. Berechne mit Hilfe eines Pythonprogrammes, die Anzahl der Räume, die innerhalb und ausserhalb des Bereichs von 30%-60% (inklusive Grenzen) liegen und gib die Resultate aus. Gib zusätzlich den Durchschnittswert der Feuchtigkeit aus. Das Resultat soll als formatierte Ausgabe zwei Nachkommastellen aufweisen.

Zu verwendende Variablen:
-   `feuchtigkeitswerte`: Liste der vom Benutzer eingegebenen Werte.
-   `anzahl_akzeptabel`, `anzahl_ausserhalb`: Anzahl der Räume mit akzeptabler bzw. ausserhalb liegender Feuchtigkeit.
-   `durchschnitt_feuchtigkeit`: Durchschnittswert der Feuchtigkeit.

```python
# Wicki Mario
# Benutzereingabe der Luftfeuchtigkeitswerte der fünf Räume.

feuchtigkeitswerte = []
for i in range(1, 6):
    wert = float(input(f'Geben Sie die Luftfeuchtigkeit für Raum {i} in Prozent ein: '))
    feuchtigkeitswerte.append(wert)

# Anzahl der Räume mit akzeptabler bzw. ausserhalb liegender Feuchtigkeit.
anzahl_akzeptabel = 0
anzahl_ausserhalb = 0

# Überprüfung der Feuchtigkeitswerte und Zählung der Räume
for wert in feuchtigkeitswerte:
    if 30 <= wert <= 60:
        anzahl_akzeptabel += 1
    else:
        anzahl_ausserhalb += 1

# Berechnung des Durchschnittswerts der Feuchtigkeit

durchschnitt_feuchtigkeit = sum(feuchtigkeitswerte) / len(feuchtigkeitswerte)

# Ausgabe der Ergebnisse

print(f'Anzahl der Räume mit akzeptabler Feuchtigkeit: {anzahl_akzeptabel}')
print(f'Anzahl der Räume mit ausserhalb liegender Feuchtigkeit: {anzahl_ausserhalb}')
print(f'Durchschnittliche Feuchtigkeit: {durchschnitt_feuchtigkeit:.2f}%')

# Kommentar: Für diese Aufgabe mit der Laufvariable i musste ich Hilfe von MathGPTPro in anspruch nehmen.
```
