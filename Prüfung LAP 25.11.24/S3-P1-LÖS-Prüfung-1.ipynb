### Aufgabe 1: Temperaturen: Durchschnitt, Maximal- und Minimalwerte
Berechnen die durchschnittliche, maximale und minimale Temperatur aus drei durch den Benutzer eingegebenen Temperaturen. Zusätzlich soll überprüft werden, ob der Durchschnittswert unter 20°C liegt, dann ist eine Meldung auszugeben: _Warnung: Durchschnittstemperatur unter dem Grenzwert!_ oder: _Durchschnittstemperatur im akzeptablen Bereich_.

Zu verwendende Variablen:
-   `temperatur1`, `temperatur2`, `temperatur3`: Die drei eingegebenen Temperaturen.
-   `durchschnitt_temperatur`: Durchschnittswert der Temperaturen.
-   `max_temperatur`, `min_temperatur`: Maximal- und Minimalwert der Temperaturen.
-   `grenzwert_durchschnitt`: Vergleichswert.

```python
# Grenzwert für Durchschnitt
grenzwert_durchschnitt = 20

# Benutzereingaben
temperatur1 = float(input("Geben Sie die erste Temperatur in °C ein: "))
temperatur2 = float(input("Geben Sie die zweite Temperatur in °C ein: "))
temperatur3 = float(input("Geben Sie die dritte Temperatur in °C ein: "))

# Durchschnitt, Max und Min berechnen
durchschnitt_temperatur = (temperatur1 + temperatur2 + temperatur3) / 3
max_temperatur = max(temperatur1, temperatur2, temperatur3)
min_temperatur = min(temperatur1, temperatur2, temperatur3)

# Ergebnisse ausgeben
print(f"Durchschnittstemperatur: {durchschnitt_temperatur:.2f} °C")
print(f"Hoechste Temperatur: {max_temperatur:.2f} °C")
print(f"Niedrigste Temperatur: {min_temperatur:.2f} °C")

# Überprüfung des Durchschnitts
if durchschnitt_temperatur < grenzwert_durchschnitt:
    print("Warnung: Durchschnittstemperatur unter dem Grenzwert!")
else:
    print("Durchschnittstemperatur im akzeptablen Bereich.")
```

### Aufgabe 3: Belueftungssysteme mit Benutzerinteraktion
Ein Belüftungssystem soll die Luftfeuchtigkeit in fünf Räumen überwachen. Der Benutzer gibt die Luftfeuchtigkeitswerte der Räume in Prozent ein. Berechne mit Hilfe eines Pythonprogrammes, die Anzahl der Räume, die innerhalb und ausserhalb des Bereichs von 30%-60% (inklusive Grenzen) liegen und gib die Resultate aus. Gib zusätzlich den Durchschnittswert der Feuchtigkeit aus. Das Resultat soll als formatierte Ausgabe zwei Nachkommastellen aufweisen.

Zu verwendende Variablen:
-   `feuchtigkeitswerte`: Liste der vom Benutzer eingegebenen Werte.
-   `anzahl_akzeptabel`, `anzahl_ausserhalb`: Anzahl der Räume mit akzeptabler bzw. ausserhalb liegender Feuchtigkeit.
-   `durchschnitt_feuchtigkeit`: Durchschnittswert der Feuchtigkeit.

```python
# Anzahl Räume festlegen
anzahl_raeume = 5

# Grenzwerte
grenzwert_min = 30
grenzwert_max = 60

# Eingabe der Feuchtigkeitswerte
feuchtigkeitswerte = []
for i in range(1, anzahl_raeume + 1):
    feuchtigkeit = float(input(f"Geben Sie die Luftfeuchtigkeit fuer Raum {i} in % ein: "))
    feuchtigkeitswerte.append(feuchtigkeit)

# Zähler für die Ergebnisse
anzahl_akzeptabel = 0
anzahl_ausserhalb = 0

# Überprüfung der Feuchtigkeitswerte
for feuchtigkeit in feuchtigkeitswerte:
    if grenzwert_min <= feuchtigkeit <= grenzwert_max:
        anzahl_akzeptabel += 1
    else:
        anzahl_ausserhalb += 1

# Durchschnitt berechnen
durchschnitt_feuchtigkeit = sum(feuchtigkeitswerte) / len(feuchtigkeitswerte)

# Ergebnisse ausgeben
print(f"Anzahl akzeptabler Raeume: {anzahl_akzeptabel}")
print(f"Anzahl Raeume ausserhalb des Bereichs: {anzahl_ausserhalb}")
print(f"Durchschnittliche Luftfeuchtigkeit: {durchschnitt_feuchtigkeit:.2f} %")
```

### Aufgabe 4: Berechnung des Energieverbrauchs
Der Benutzer soll die Energieverbräuche von zehn Geräten eingeben können. Berechne den Gesamtverbrauch, den Durchschnittsverbrauch und gib an, ob der Durchschnittsverbrauch einen Grenzwert von 3 kWh überschreitet. Beim Überschreiten soll die Meldung _Warnung: Durchschnittsverbrauch überschreitet den Grenzwert!_ und sonst: _Durchschnittsverbrauch im akzeptablen Bereich._ ausgegeben werden.

Zu verwendende Variablen:
-   `energieverbraeuche`: Liste der eingegebenen Energieverbräuche.
-   `gesamtverbrauch`: Gesamter Energieverbrauch.
-   `durchschnitt_verbrauch`: Durchschnittlicher Energieverbrauch.
-   `grenzwert_durchschnitt`: Vergleichswert für den Durchschnitt.

```python
# Anzahl Geräte
geraete = 10

# Grenzwert
grenzwert_durchschnitt = 3

# Eingabe der Energieverbräuche
energieverbraeuche = []
for i in range(1, geraete + 1):
    verbrauch = float(input(f"Geben Sie den Energieverbrauch von Geraet {i} in kWh ein: "))
    energieverbraeuche.append(verbrauch)
    
# Berechnungen
gesamtverbrauch = sum(energieverbraeuche)
durchschnitt_verbrauch = gesamtverbrauch / len(energieverbraeuche)

# Ergebnisse ausgeben
print(f"Gesamter Energieverbrauch: {gesamtverbrauch:.2f} kWh")
print(f"Durchschnittlicher Energieverbrauch: {durchschnitt_verbrauch:.2f} kWh")

# Überprüfung des Durchschnitts
if durchschnitt_verbrauch > grenzwert_durchschnitt:
    print("Warnung: Durchschnittsverbrauch überschreitet den Grenzwert!")
else:
    print("Durchschnittsverbrauch liegt innerhalb des akzeptablen Bereichs")
```

### Aufgabe 5: Temperaturkonvertierung mit Benutzerinteraktion
Der Benutzer gibt fünf Temperaturen in Celsius ein. Verwende hierzu eine geeignete Schleife. Konvertiere diese anschliessend in Fahrenheit mit Hilfe dieser [Formel](https://de.wikipedia.org/wiki/Grad_Celsius) und einer Funktion _konvertiere_zu_fahrenheit_. Berechne anschliessend den Durchschnitt der Fahrenheit-Werte auf zwei Nachkommastellen genau und gib die Einzelwerte und das Resultat aus.

Zu verwendende Variablen:
-   `temperaturen_celsius`: Liste der eingegebenen Temperaturen in Celsius.
-   `temperaturen_fahrenheit`: Liste der Temperaturen in Fahrenheit.
-   `durchschnitt_fahrenheit`: Durchschnitt der Temperaturen in Fahrenheit.

```python
# Temperaturwerte
anzahl = 5

# Funktion zur Umrechnung in Fahrenheit
def konvertiere_zu_fahrenheit(temperatur_celsius):
    return temperatur_celsius * 1.8 + 32

# Eingabe der Temperaturen in Celsius
temperaturen_celsius = []
for i in range(1, anzahl + 1):
    temperatur = float(input(f"Geben Sie die Temperatur {i} in Grad Celsius ein: "))
    temperaturen_celsius.append(temperatur)

# Umrechnen in Fahrenheit
temperaturen_fahrenheit = []
for temp in temperaturen_celsius:
    temperaturen_fahrenheit.append(konvertiere_zu_fahrenheit(temp))

# Durchschnitt berechnen
durchschnitt_fahrenheit = sum(temperaturen_fahrenheit) / len(temperaturen_fahrenheit)

# Ergebnisse ausgeben
print("Temperaturen in Celsius:", temperaturen_celsius)
print("Temperaturen in Fahrenheit:", temperaturen_fahrenheit)
print(f"Durchschnittstemperatur in Fahrenheit: {round(durchschnitt_fahrenheit,2)} Grad Fahrenheit")
```

### Aufgabe 6: Berechnung von CO₂-Emissionen pro Gerät
Der Benutzer gibt die Energieverbräuche von fünf Geräten einzeln in kWh ein. Verwende dazu eine Schleife. Berechne die CO₂-Emissionen jedes Geräts und die Gesamtemissionen bei einem vordefinierten Emissionsfaktor von 0.4 kg CO₂/kWh. Gib jeweils die einzelnen Werte aus und implementiere für die Berechnung der einzelnen Emissionen eine Funktion _berechne_emissionen_ und runde die Resultatwerte auf zwei Nachkommastellen.

Zu verwendende Variablen:

-   `energieverbraeuche`: Liste der eingegebenen Energieverbräuche.
-   `emissionsfaktor` = 0.4: Emissionsfaktor.
-   `emissionen_pro_geraet`: Liste der berechneten Emissionen pro Gerät.
-   `gesamt_emissionen`: Gesamte CO₂-Emissionen.

```python
# Anzahl Geräte
anzahl = 5

# Emissionsfaktor definieren
emissionsfaktor = 0.4

# Funktion zur Berechnung der Emissionen
def berechne_emissionen(energieverbrauch, emissionsfaktor):
    return energieverbrauch * emissionsfaktor

# Eingabe der Energieverbräuche
energieverbraeuche = []
for i in range(1, anzahl + 1):
    verbrauch = float(input(f"Geben Sie den Energieverbrauch von Geraet {i} in kWh ein: "))
    energieverbraeuche.append(verbrauch)

# Emissionen pro Geraet berechnen
emissionen_pro_geraet = []
for verbrauch in energieverbraeuche:
    emissionen_pro_geraet.append(berechne_emissionen(verbrauch, emissionsfaktor))

# Gesamtemissionen berechnen
gesamt_emissionen = sum(emissionen_pro_geraet)

# Ergebnisse ausgeben
for i in range(anzahl):
    print(f"Geraet {i + 1}: Eingabe: {energieverbraeuche[i]} Ausgabe: {emissionen_pro_geraet[i]:.2f}")
    
print(f"Gesamte CO₂-Emissionen: {round(gesamt_emissionen, 2)} kg")
```

### Aufgabe 7: Einfache Liftsteuerung
Schreibe ein Programm, das die Bewegungen eines Lifts simuliert. Du kannst Zielstockwerke angeben, die der Lift ansteuern soll. Das Programm zeigt Dir an, ob sich der Lift nach oben oder unten bewegt, oder ob er stehen bleibt. Ausserdem zählt es, wie oft der Lift nach oben und unten gefahren ist, und speichert die angefahrenen Stockwerke. Das Programm endet, wenn Du eine spezielle Eingabe machst.


Beschreibung der Anforderungen:
1.  Liftgrenzen und Startbedingungen:
    -   Der Lift bewegt sich zwischen den Stockwerken 0 (Minimum) und 10 (Maximum).
    -   Am Anfang steht der Lift auf Stockwerk 0.
2.  Eingabe der Zielstockwerke:
    -   Gib ein Zielstockwerk ein, das der Lift ansteuern soll.
    -   Mit der Eingabe -1 kannst Du das Programm beenden.
    -   Falls Du ein ungültiges Stockwerk angibst (ausserhalb des Bereichs), weist Dich das Programm darauf hin und fordert Dich zur erneuten Eingabe auf.
3.  Bewegung des Lifts:
    -   Wenn das Zielstockwerk höher als das aktuelle Stockwerk ist, fährt der Lift nach oben.
    -   Ist das Zielstockwerk niedriger, fährt der Lift nach unten.
    -   Wenn der Lift schon auf dem Zielstockwerk steht, bleibt er dort.
4.  Protokollierung der Fahrten:
    -   Das Programm speichert alle angefahrenen Zielstockwerke in einer Liste.
    -   Es zählt, wie oft der Lift nach oben und nach unten gefahren ist.
5.  Ausgabe:
    -   Während des Programms informiert Dich der Lift darüber, ob er sich nach oben oder unten bewegt, oder ob er stehen bleibt.
    -   Am Ende zeigt Dir das Programm eine Zusammenfassung:
        -   Eine Liste der angefahrenen Stockwerke.
        -   Die Anzahl der Fahrten nach oben und nach unten.
6.  Robustheit der Eingabe:
    -   Das Programm überprüft Deine Eingaben, damit nur gültige Werte verarbeitet werden.
    -   Ungültige Eingaben werden ignoriert und zur erneuten Eingabe aufgefordert.

```python
# 'konstante' Variablen
max_stockwerke = 10
min_stockwerke = 0

# Startwerte
aktuelle_position = 0
fahrten_hoch = 0
fahrten_runter = 0
angefahrene_stockwerke = []
gueltige_position = True

print("\n--- Liftsteuerung gestartet ---")
print("Gib die Zielstockwerke ein (zwischen 0 und 10). Beenden mit -1.")

while True:
    # Zielstockwerk eingeben
    ziel_stockwerk = int(input(f"Aktuelle Position: {aktuelle_position}. Geben Sie das Zielstockwerk ein: "))
    
    # Beenden der Eingabe
    if ziel_stockwerk == -1:
        break

    # Überprüfung auf gültigen Bereich und Bewegung des Liftes
    if (ziel_stockwerk < min_stockwerke) or (ziel_stockwerk > max_stockwerke):
        print(f"Ungültiges Stockwerk: {ziel_stockwerk}. Gib ein Stockwerk zwischen {min_stockwerke} und {max_stockwerke} ein.")
        gueltige_position = False
    elif ziel_stockwerk > aktuelle_position:
        print(f"Lift fährt nach oben von {aktuelle_position} zu {ziel_stockwerk}.")
        fahrten_hoch += 1
        gueltige_position = True
    elif ziel_stockwerk < aktuelle_position:
        print(f"Lift fährt nach unten von {aktuelle_position} zu {ziel_stockwerk}.")
        fahrten_runter += 1
        gueltige_position = True
    else:
        print(f"Lift bleibt auf dem aktuellen Stockwerk {aktuelle_position}.")
        gueltige_position = True
    
    # Update der aktuellen Position und Speicherung
    if gueltige_position:
        aktuelle_position = ziel_stockwerk
        angefahrene_stockwerke.append(ziel_stockwerk)

# Zusammenfassung der Ergebnisse
print("\n--- Liftsteuerung beendet ---")
print(f"Angefahrene Stockwerke: {angefahrene_stockwerke}")
print(f"Fahrten nach oben: {fahrten_hoch}")
print(f"Fahrten nach unten: {fahrten_runter}")
```
