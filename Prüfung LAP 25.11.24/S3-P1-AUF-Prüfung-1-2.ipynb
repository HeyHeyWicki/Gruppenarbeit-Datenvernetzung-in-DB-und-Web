### Aufgabe 2: Heizsystemsteuerung mit variablen Grenzwerten
Das Heizsystem soll aktiviert werden, wenn die aktuelle Raumtemperatur unterhalb eines durch den Benutzer eingegebenen Grenzwerts liegt. Zusätzliche Benutzerabfrage: Ist das Heizsystem im Wartungsmodus (ja/nein)? Falls ja, soll die Heizung nicht aktiviert werden, falls nein soll sie gemäss den beschriebenen Bedingungen aktiviert werden. Erstelle ein entsprechendes Pythonprogramm und gib die einzelnen Meldungen aus.

Zu verwendende Variablen:
-   `aktuelle_temperatur`: Vom Benutzer eingegebene Temperatur.
-   `grenzwert_heizung`: Vom Benutzer eingegebener Grenzwert.
-   `wartungsmodus`: Status des Wartungsmodus (ja/nein).

```python
# Wicki Mario
# Vom Benutzer einzugebende Temperatur.

Aktuelle_Temperatur = float(input('Geben sie die aktuelle Raumtemperatur ein: '))
Grenzwert_Heizung = float(input('Geben sie den Grenzwert für die Heizung ein: '))

# Benutzerabfrage Wartungsmodus.

Wartungsmodus = input('Wartungsmodus(ja/nein)?: ').strip().lower()

# Überprüfung der Bedingungen und Ausgabe der entsprechenden Meldungen

if Wartungsmodus == 'ja':
    print('Heizung ist im Wartungsmodus und wird nicht aktiviert.')
elif Aktuelle_Temperatur < Grenzwert_Heizung:
    print('Aktuelle Temperatur ist unterhalb des Grenzwerts. Heizung wird aktiviert.')
else:
    print('Aktuelle Temperatur ist oberhalb des Grenzwerts. Heizung wird nicht aktiviert.')

# Kommentar: Um die Aufgabe mit dem Wartungsmodus zu lösen (.strip und .lower) sowie mit dem Befehl elif musste ich Hilfe von MathGPTPro nehmen.
```
