### Aufgabe 1: Temperaturen: Durchschnitt, Maximal- und Minimalwerte
Berechnen die durchschnittliche, maximale und minimale Temperatur aus drei durch den Benutzer eingegebenen Temperaturen. Zusätzlich soll überprüft werden, ob der Durchschnittswert unter 20°C liegt, dann ist eine Meldung auszugeben: _Warnung: Durchschnittstemperatur unter dem Grenzwert!_ oder: _Durchschnittstemperatur im akzeptablen Bereich_.

Zu verwendende Variablen:
-   `temperatur1`, `temperatur2`, `temperatur3`: Die drei eingegebenen Temperaturen.
-   `durchschnitt_temperatur`: Durchschnittswert der Temperaturen.
-   `max_temperatur`, `min_temperatur`: Maximal- und Minimalwert der Temperaturen.
-   `grenzwert_durchschnitt`: Vergleichswert.

```python
# Wicki Mario
# Drei einzugebenden Temperaturen durch den Benutzer.

Temperatur1 = float(input('Geben sie die erste Temperatur ein: '))
Tempreatur2 = float(input('geben sie die zweite Temeperarur ein: '))
Temperatur3 = float(input('geben sie die dritte Temperatur ein: '))

# Berechnung des Durchschnittswerts.

Durchschnitt_Temperatur = (Temperatur1 + Temperatur3 + Temperatur3) /3

# Maximal- und Minimalwert der Temperatur.

max_Temperatur = max(Temperatur1, Tempreatur2, Temperatur3)
min_Temperatur = min(Temperatur1, Tempreatur2, Temperatur3)


# Ausgabe der Ergebnisse.

print(f'Durchschnittstemperatur: {Durchschnitt_Temperatur:.2f}°C')
print(f'Maximale Temperatur: {max_Temperatur:.2f}°C')
print(f'Minimale Temperatur: {min_Temperatur:.2f}°C')

# Warnung: Durchschnittstemperatur unter dem Grenzweert! oder: Durchschnittstemperatur im akzeptablen Bereich.

grenzwert_durchschnitt = 20.0

if Durchschnitt_Temperatur < grenzwert_durchschnitt:
    print('Warnung: Durchschnittstemperatur unter dem Grenzwert!')
else:
    print('Durchschnittstemperatur im akzeptablen Bereich.')

# Kommentar: Bei der Ausgabe der Ergebnisse sowie beim Grenzwert Durchschnitt habe ich Hilfe von MathGPTpro geholt.

```
