### Aufgabe 4: Berechnung des Energieverbrauchs
Der Benutzer soll die Energieverbräuche von zehn Geräten eingeben können. Berechne den Gesamtverbrauch, den Durchschnittsverbrauch und gib an, ob der Durchschnittsverbrauch einen Grenzwert von 3 kWh überschreitet. Beim Überschreiten soll die Meldung _Warnung: Durchschnittsverbrauch überschreitet den Grenzwert!_ und sonst: _Durchschnittsverbrauch im akzeptablen Bereich._ ausgegeben werden.

Zu verwendende Variablen:
-   `energieverbraeuche`: Liste der eingegebenen Energieverbräuche.
-   `gesamtverbrauch`: Gesamter Energieverbrauch.
-   `durchschnitt_verbrauch`: Durchschnittlicher Energieverbrauch.
-   `grenzwert_durchschnitt`: Vergleichswert für den Durchschnitt.

```python
# Wicki Mario
# Benutzereingabe für die Energieverbräuche der zehn Geräte

energieverbraeuche = []
for i in range(1, 11):
    verbrauch = float(input(f'Geben Sie den Energieverbrauch für Gerät {i} in kWh ein: '))
    energieverbraeuche.append(verbrauch)

# Berechnung des Gesamtverbrauchs
gesamtverbrauch = sum(energieverbraeuche)

# Berechnung des Durchschnittsverbrauchs
durchschnitt_verbrauch = gesamtverbrauch / len(energieverbraeuche)
# Ausgabe der Ergebnisse

print(f'Gesamtverbrauch: {gesamtverbrauch:.2f} kWh')
print(f'Durchschnittsverbrauch: {durchschnitt_verbrauch:.2f} kWh')

# Überprüfung, ob der Durchschnittsverbrauch den Grenzwert überschreitet
grenzwert_durchschnitt = 3.0
if durchschnitt_verbrauch > grenzwert_durchschnitt:
    print('Warnung: Durchschnittsverbrauch überschreitet den Grenzwert!')
else:
    print('Durchschnittsverbrauch im akzeptablen Bereich.')

# Kommentar: Für dise >Aufgabe mit der Laufvariable i musste Hilfe von MathGPTPro in Anspruch nehmen.
```
