### Aufgabe 7: Einfache Liftsteuerung
Schreibe ein Programm, das die Bewegungen eines Lifts simuliert. Du kannst Zielstockwerke angeben, die der Lift ansteuern soll. Das Programm zeigt Dir an, ob sich der Lift nach oben oder unten bewegt, oder ob er stehen bleibt. Ausserdem zählt es, wie oft der Lift nach oben und unten gefahren ist, und speichert die angefahrenen Stockwerke. Das Programm endet, wenn Du eine spezielle Eingabe machst.


```python
Beschreibung der Anforderungen:
1.  Liftgrenzen und Startbedingungen:
    -   Der Lift bewegt sich zwischen den Stockwerken 0 (Minimum) und 10 (Maximum).
    -   Am Anfang steht der Lift auf Stockwerk 0.
2.  Eingabe der Zielstockwerke:
    -   Gib ein Zielstockwerk ein, das der Lift ansteuern soll.
    -   Mit der Eingabe -1 kannst Du das Programm beenden.
    -   Falls Du ein ungültiges Stockwerk angibst (ausserhalb des Bereichs), weist Dich das Programm darauf hin und fordert Dich zur erneuten Eingabe auf.
3.  Bewegung des Lifts:
    -   Wenn das Zielstockwerk höher als das aktuelle Stockwerk ist, fährt der Lift nach oben.
    -   Ist das Zielstockwerk niedriger, fährt der Lift nach unten.
    -   Wenn der Lift schon auf dem Zielstockwerk steht, bleibt er dort.
4.  Protokollierung der Fahrten:
    -   Das Programm speichert alle angefahrenen Zielstockwerke in einer Liste.
    -   Es zählt, wie oft der Lift nach oben und nach unten gefahren ist.
5.  Ausgabe:
    -   Während des Programms informiert Dich der Lift darüber, ob er sich nach oben oder unten bewegt, oder ob er stehen bleibt.
    -   Am Ende zeigt Dir das Programm eine Zusammenfassung:
        -   Eine Liste der angefahrenen Stockwerke.
        -   Die Anzahl der Fahrten nach oben und nach unten.
6.  Robustheit der Eingabe:
    -   Das Programm überprüft Deine Eingaben, damit nur gültige Werte verarbeitet werden.
    -   Ungültige Eingaben werden ignoriert und zur erneuten Eingabe aufgefordert.
```

```python
# Wicki Mario

def lift_simulation():
    # Initialisierung der Variablen
    aktuelles_stockwerk = 0
    angefahrene_stockwerke = []
    fahrten_nach_oben = 0
    fahrten_nach_unten = 0
    while True:
        try:
            # Eingabe des Zielstockwerks
            ziel_stockwerk = int(input('Geben Sie das Zielstockwerk ein (0-10) oder -1 zum Beenden: '))
            # Beenden des Programms
            if ziel_stockwerk == -1:
                break
            # Überprüfung der Gültigkeit des Zielstockwerks
            if ziel_stockwerk < 0 or ziel_stockwerk > 10:
                print('Ungültiges Stockwerk. Bitte geben Sie ein Stockwerk zwischen 0 und 10 ein.')
                continue
            # Bewegung des Lifts
            if ziel_stockwerk > aktuelles_stockwerk:
                print(f'Der Lift bewegt sich von Stockwerk {aktuelles_stockwerk} nach oben zu Stockwerk {ziel_stockwerk}.')
                fahrten_nach_oben += 1
            elif ziel_stockwerk < aktuelles_stockwerk:
                print(f'Der Lift bewegt sich von Stockwerk {aktuelles_stockwerk} nach unten zu Stockwerk {ziel_stockwerk}.')
                fahrten_nach_unten += 1
            else:
                print(f'Der Lift bleibt auf Stockwerk {aktuelles_stockwerk}.')
            # Aktualisierung des aktuellen Stockwerks und Protokollierung
            aktuelles_stockwerk = ziel_stockwerk
            angefahrene_stockwerke.append(ziel_stockwerk)
        except ValueError:
            print('Ungültige Eingabe. Bitte geben Sie eine Zahl ein.')
    # Ausgabe der Zusammenfassung
    print('\nZusammenfassung:')
    print(f'Angefahrene Stockwerke: {angefahrene_stockwerke}')
    print(f'Anzahl der Fahrten nach oben: {fahrten_nach_oben}')
    print(f'Anzahl der Fahrten nach unten: {fahrten_nach_unten}')
# Start der Lift-Simulation
lift_simulation()

# Kommentar: Diese Aufgabe habe ich vollumfänglich mit MathGPTPro gelöst.
```
