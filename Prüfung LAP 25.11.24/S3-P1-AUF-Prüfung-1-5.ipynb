### Aufgabe 5: Temperaturkonvertierung mit Benutzerinteraktion
Der Benutzer gibt fünf Temperaturen in Celsius ein. Verwende hierzu eine geeignete Schleife. Konvertiere diese anschliessend in Fahrenheit mit Hilfe dieser [Formel](https://de.wikipedia.org/wiki/Grad_Celsius) und einer Funktion _konvertiere_zu_fahrenheit_. Berechne anschliessend den Durchschnitt der Fahrenheit-Werte auf zwei Nachkommastellen genau und gib die Einzelwerte und das Resultat aus.

Zu verwendende Variablen:
-   `temperaturen_celsius`: Liste der eingegebenen Temperaturen in Celsius.
-   `temperaturen_fahrenheit`: Liste der Temperaturen in Fahrenheit.
-   `durchschnitt_fahrenheit`: Durchschnitt der Temperaturen in Fahrenheit.

```python
# Wicki Mario
# Funktion zur Konvertierung von Celsius zu Fahrenheit
def konvertiere_zu_fahrenheit(celsius):
    return celsius * 9/5 + 32

# Benutzereingabe für die fünf Temperaturen in Celsius
temperaturen_celsius = []
for i in range(1, 6):
    temperatur = float(input(f'Geben Sie die Temperatur {i} in Celsius ein: '))
    temperaturen_celsius.append(temperatur)

# Konvertierung der Temperaturen in Fahrenheit
temperaturen_fahrenheit = [konvertiere_zu_fahrenheit(temp) for temp in temperaturen_celsius]

# Berechnung des Durchschnitts der Fahrenheit-Werte
durchschnitt_fahrenheit = sum(temperaturen_fahrenheit) / len(temperaturen_fahrenheit)

# Ausgabe der Einzelwerte und des Durchschnitts
print('Temperaturen in Fahrenheit:')
for i, temp in enumerate(temperaturen_fahrenheit, start=1):
    print(f'Temperatur {i}: {temp:.2f}°F')
print(f'Durchschnittliche Temperatur in Fahrenheit: {durchschnitt_fahrenheit:.2f}°F')

# Kommentar: Diese Aufgabe habe ich vollumfänglich mit MathGPTPro gelöst.
```
