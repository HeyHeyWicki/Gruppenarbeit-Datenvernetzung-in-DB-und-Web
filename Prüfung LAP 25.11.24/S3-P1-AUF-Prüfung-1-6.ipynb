### Aufgabe 6: Berechnung von CO₂-Emissionen pro Gerät
Der Benutzer gibt die Energieverbräuche von fünf Geräten einzeln in kWh ein. Verwende dazu eine Schleife. Berechne die CO₂-Emissionen jedes Geräts und die Gesamtemissionen bei einem vordefinierten Emissionsfaktor von 0.4 kg CO₂/kWh. Gib jeweils die einzelnen Werte aus und implementiere für die Berechnung der einzelnen Emissionen eine Funktion _berechne_emissionen_ und runde die Resultatwerte auf zwei Nachkommastellen.

Zu verwendende Variablen:

-   `energieverbraeuche`: Liste der eingegebenen Energieverbräuche.
-   `emissionsfaktor` = 0.4: Emissionsfaktor.
-   `emissionen_pro_geraet`: Liste der berechneten Emissionen pro Gerät.
-   `gesamt_emissionen`: Gesamte CO₂-Emissionen.

```python
# Wicki Mario
# Funktion zur Berechnung der CO2-Emissionen
def berechne_emissionen(verbrauch, emissionsfaktor):
    return verbrauch * emissionsfaktor

# Vordefinierter Emissionsfaktor
emissionsfaktor = 0.4

# Benutzereingabe für die Energieverbräuche der fünf Geräte
energieverbraeuche = []
for i in range(1, 6):
    verbrauch = float(input(f'Geben Sie den Energieverbrauch für Gerät {i} in kWh ein: '))
    energieverbraeuche.append(verbrauch)

# Berechnung der CO2-Emissionen pro Gerät
emissionen_pro_geraet = [berechne_emissionen(verbrauch, emissionsfaktor) for verbrauch in energieverbraeuche]

# Berechnung der gesamten CO2-Emissionen
gesamt_emissionen = sum(emissionen_pro_geraet)

# Ausgabe der einzelnen Emissionen und der Gesamtemissionen
print('CO2-Emissionen pro Gerät:')
for i, emission in enumerate(emissionen_pro_geraet, start=1):
    print(f'Gerät {i}: {emission:.2f} kg CO2')
print(f'Gesamte CO2-Emissionen: {gesamt_emissionen:.2f} kg CO2')

# Kommentar: Diese Aufgabe habe ich vollumfänglich mit MathGPTPro gelöst.
```
