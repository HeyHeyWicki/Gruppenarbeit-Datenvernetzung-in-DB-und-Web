## Aufgabe 3 - Sortierung mit dem Quicksort-Algorithmus
### a) Quicksort (rekursiv) aufsteigend
Erstelle mit Hilfe des Buches "Algorithmen kapieren", Kapitel "Quicksort" oder einer KI ein Pythonprogramm, welches eine Liste mit unsortierten Zahlen entgegennimmt und diese anschliessend aufsteigend sortiert ausgibt.
#### mit List-Comprehension:

```python
def quicksort(liste):
    # Basisfall
    if len(liste) <= 1:
        return liste
    else:
        pivot    = liste[0]
        kleiner  = [i for i in liste[1:] if i <= pivot]
        groesser = [i for i in liste[1:] if i > pivot]
        # Rekursionsfall
        return quicksort(kleiner) + [pivot] + quicksort(groesser)
    
print(quicksort([10, 5, 2, 3]))
```

#### ohne List-Comprehension:

```python
def quicksort(liste):
    # Basisfall
    if len(liste) <= 1:
        return liste
    else:
        pivot = liste[0]
        
        kleiner = []
        groesser = []
        
        # Iteriere über die Elemente ab dem zweiten
        for i in liste[1:]:
            if i <= pivot:
                kleiner.append(i)
            else:
                groesser.append(i)
        # Rekursionsfall
        return quicksort(kleiner) + [pivot] + quicksort(groesser)

print(quicksort([10, 5, 2, 3]))
```

### b) Quicksort (rekursiv) absteigend
Erstelle mit Hilfe des Buches "Algorithmen kapieren", Kapitel "Quicksort" oder einer KI ein Pythonprogramm, welches eine Liste mit unsortierten Zahlen entgegennimmt und diese anschliessend absteigend sortiert ausgibt.

Um die Sortierreihenfolge in absteigender Reihenfolge zu ändern, müssen wir lediglich die Vergleichsoperationen in der quicksort-Funktion umdrehen. Elemente, die grösser als das Pivot-Element sind, sollen zuerst verarbeitet werden.

```python
def quicksort(liste):
    if len(liste) <= 1:
        return liste
    else:
        pivot    = liste[0]
        kleiner  = [i for i in liste[1:] if i <= pivot]
        groesser = [i for i in liste[1:] if i > pivot]
        return quicksort(groesser) + [pivot] + quicksort(kleiner)

print(quicksort([10, 5, 2, 3]))
```

## Aufgabe 4 - Selection-Sort
Erstelle mit Hilfe des Buches "Algorithmen kapieren", Kapitel "Selectionsort" oder mit Hilfe einer KI ein Pythonprogramm, welches eine Liste mit unsortierten Zahlen entgegennimmt und diese anschliessend aufsteigend sortiert ausgibt.

### a) Variante 1: iterativ

```python
personen = [10, 5, 2, 3]

def selection_sort(liste):
    sortierte_liste = []
    while liste:
        # Finde das kleinste Element in der Liste
        min_index = 0
        for i in range(1, len(liste)):
            if liste[i] < liste[min_index]:
                min_index = i
        # füge das kleinste Element zur sortierten Liste hinzu (append)
        # und entferne es aus der Ausgangsliste (pop)
        sortierte_liste.append(liste.pop(min_index))
    return sortierte_liste

# Sortiere die Anzahl Personen in den einzelnen Wohnungen
sortierte_liste = selection_sort(personen)

# Ausgabe
print("Sortierte Liste:", sortierte_liste)
```

### b) Variante 2: rekursiv

```python
personen = [10, 5, 2, 3]

def selection_sort_rekursiv(liste, start=0):
    # Basisfall: Wenn der Startindex das Ende der Liste erreicht hat, ist die Liste sortiert
    if start >= len(liste) - 1:
        return liste
    
    # Finde den Index des kleinsten Elements im unsortierten Teil der Liste
    min_index = start
    for i in range(start + 1, len(liste)):
        if liste[i] < liste[min_index]:
            min_index = i
    
    # Tausche das aktuelle Element mit dem kleinsten Element
    liste[start], liste[min_index] = liste[min_index], liste[start]
    
    # Rekursiver Aufruf für den Rest der Liste
    return selection_sort_rekursiv(liste, start + 1)

# Sortiere die Anzahl Personen in den einzelnen Wohnungen
sortierte_liste = selection_sort_rekursiv(personen)

# Ausgabe
print("Sortierte Liste:", sortierte_liste)

```
