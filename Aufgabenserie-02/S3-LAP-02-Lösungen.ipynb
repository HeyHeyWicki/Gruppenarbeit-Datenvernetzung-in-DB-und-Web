## Aufgabe 5
Gegeben ist die folgende Zeichenfolge: "bla bla [wichtig] mehr bla bla...". Extrahiere daraus den Teil, welcher zwischen den beiden eckigen Klammern steht.

## Aufgabe 6
Wir wollen mit Python einige Brüche addieren, subtrahieren, multiplizieren und dividieren. dazu können wir die Befehle aus dem Modul _fractions_ importieren. Gib dazu zuoberst in deinem Programm die Anweisung:
```python 
from fractions import Fractions
```
ein. Erstelle selber ein paar Beispiele und prüfe nach, ob die Resultate stimmen.


#### a) ein paar Beispiele
Wir addieren, subtrahieren, multiplizieren und dividieren die Brüche.

````python
from fractions import Fraction

print(Fraction(5, 6) + Fraction(1, 3))      # 5/6 + 1/3
print(Fraction(1, 3) - Fraction(2, 3))      # 1/3 - 2/3
print(Fraction(3, 4) * Fraction(2, 3))      # 3/4 * 2/3
print(Fraction(2, 7) / Fraction(2, 9))      # 2/7 / 2/9
````

#### b) Erweiterung
Erweitere dein Programm so, dass der Benutzer eingeben kann, welche Brüche er verwenden will und wie er mit diesen operiert, das heisst, ob er diese addieren, subtrahieren, multiplizieren oder dividieren will. Verwende dazu eine Verzweigung / Selektion verwenden:
```python 
if Bedingung:
    dann...
elif:
    sonst Bedinung...
else:
    sonst...
```
Denke daran, dass du mit den eingegebenen "Werten" rechnen willst!

````python
from fractions import Fraction

zaehler_1 = int(input('Gib den Zähler des ersten Bruches ein:'))
nenner_1 = int(input('Gib den Nenner des ersten Bruches ein:'))

zaehler_2 = int(input('Gib den Zähler des zweiten Bruches ein:'))
nenner_2 = int(input('Gib den Nenner des zweiten Bruches ein:'))

print('Addition: a')
print('Subtraktion: s')
print('Multiplikation: m')
print('Division: d')
operation = input('Wähle die Operation aus:')

if operation == 'a':
    print('das Resultat lautet:', Fraction(zaehler_1, nenner_1) + Fraction(zaehler_2, nenner_2))
elif operation == 's':
    print('das Resultat lautet:', Fraction(zaehler_1, nenner_1) - Fraction(zaehler_2, nenner_2))
elif operation == 'm':
    print('das Resultat lautet:', Fraction(zaehler_1, nenner_1) * Fraction(zaehler_2, nenner_2))
elif operation == 'd':
    print('das Resultat lautet:', Fraction(zaehler_1, nenner_1) / Fraction(zaehler_2, nenner_2))
else:
    print("Falsche Auswahl!")
````

## Aufgabe 7
Erstelle ein Pythonprogramm, welches die Grund-, Mantel-, Oberfläche und das Volumen eines geraden Kreiszylinders berechnet. Du sollst den Radius (r) und die Höhe (h) eingeben können. Anschliessend werden die beiden gesuchten Grössen durch das Programm berechnet und ausgegeben. Damit du mit Pi rechnen kannst, musst du die Bibliothek __math__ importieren.

````python
import math as math

r = float(input("Gib den Radius (r) der Grund-/Deck-Fläche ein:"))
h = float(input("Gib die Höhe (h) des Kreiszylinders ein:"))

# Grund-/Deck-Fläche berechnen
grund_flaeche = r * r * math.pi

# Mantelfläche berechnen
umfang = 2 * r * math.pi
mantel_flaeche = umfang * h

# Oberfläche berechnen: 2 * Grundfläche + Mantelfläche
ober_flaeche = 2 * grund_flaeche + mantel_flaeche

# Volumen berechnen: Grundfläche * Höhe
volumen = grund_flaeche * h

# Ausgabe der Eingangs- und Ausgangsgrösse
print("Radius (r) der Grundfläche:", r)
print("Höhe (h) des Kreiszylinders:", h)
# gerundet auf 3 Nachkommastellen
print("Die Grundfläche des Kreiszylinders beträgt:", "%.3f " % grund_flaeche)
print("Die Mantelfläche des Kreiszylinders beträgt:", "%.3f " % mantel_flaeche)
print("Die Oberfläche des Kreiszylinders beträgt:", "%.3f " % ober_flaeche)
print("Die Oberfläche des Kreiszylinders beträgt:", "%.3f " % volumen)
````

## Aufgabe 8
Erstelle ein Pythonprogramm, welches dich nach einer natürlichen Zahl fragt und anschliessend ausgibt, ob diese Zahl __gerade__ oder __ungerade__ ist. Dabei musst du eine Verzweigung / Selektion verwenden:
```python 
if Bedingung:
    dann...
else:
    sonst...
```

#### a) ohne "divmod"-Operator

````python
zahl = int(input("Gib eine natürliche Zahl ein: "))
if (zahl // 2) * 2 == zahl:
    print(zahl, "ist gerade")
else:
    print(zahl, "ist ungerade")
````

#### b) mit "divmod"-Operator

````python
zahl = int(input("Gib eine natürliche Zahl ein: "))

# Das Resultat von divmod ist ein Tupel mit dem ganzzahligen Wert und dem Rest bei der Division
resultat = divmod(zahl, 2)

if resultat[1] == 0:
    print(zahl, " ist gerade")
else:
    print(zahl, " ist ungerade")
````
