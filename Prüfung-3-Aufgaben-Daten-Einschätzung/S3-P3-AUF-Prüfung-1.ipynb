## Aufgabe 1
Von einzelnen Räumen in einem Gebäude, sind die täglich und zufällig gemessenen Raumtemperaturen, in Grad Celsius in einer Datei gespeichert worden. Man wollte damit eine Qualitätskontrolle durchführen und feststellen, wie oft die Vorgabetemperatur von 18 Grad Celsius (plus oder minus 1 Grad Celsius als Abweichungstoleranz) unter oder überschritten worden ist. Leider ist die Datei von einem Raum unabsichtlich gelöscht worden. 
### a) Daten generieren
Damit wir trotzdem ein Gesamtbild darstellen können musst du für diesen Raum die Daten nach den folgenden Vorgaben generieren:
1. Es sind Tagesdaten für das Jahr 2024 zu generieren.
2. Du kannst von normalverteilten Daten ausgehen.
3. Das arithmetische Mittel beträgt 18 Grad Celsius.
4. Die Abweichung (Standardabweichung) von diesem Mittelwert beträgt 2 Grad.
5. Als Spaltentrennzeichen soll das Zeichen ";" verwendet werden.
6. Anzeigen der ersten fünf gespeicherten Datensätze.

Die so generierten Daten sollen in der Datei _daten-p3-aus-1.txt_ gespeichert werden. Es sind die folgenden Spalten mit den Daten aufzulisten:
- Datum
- Tagestemperatur

```python
## Kommentar Wicki Mario...
import numpy as np
import pandas as pd
#Generiere Datumsspalte für das Jahr 2024
dates = pd.date_range(start='2024-01-01', end='2024-12-31', freq='D')
#Generiere Temperaturwerte aus N(18, 2^2)
temperatures = np.random.normal(loc=18, scale=2, size=len(dates))
#Erstelle DataFrame
df = pd.DataFrame({'Datum': dates.strftime('%Y-%m-%d'), 'Tagestemperatur': np.round(temperatures,2)})
#Speichere als CSV mit Semikolon als Trenner
df.to_csv('daten-p3-aus-1.txt', sep=';', index=False)
#Ausgabe der ersten 5 Datensätze:
print(df.head())
```

### b) Daten visualisieren
Stelle die in a) generierten Daten möglichst in einem geeigneten Diagramm dar und ergänze zur Visualisierung noch ein Boxplot. Zudem soll das arithmetische Mittel im Boxplot ersichtlich sein. Anschliessend sollen die wichtigsten statistischen Kennzahlen durch den Befehl `describe()` angezeigt werden.

Verwende die Bibliothek `seaborn` und daraus den Befehl `scatterplot()` um die Daten entsprechend visualisieren zu können. 

Damit diese auf der horizontalen Achse einigermassen dargestellt werden, kannst du den Parameter `parse_dates=['Datum']` beim Einlesen mit der `pandas`-Bibliothek verwenden.

```python
## hier deinen Code eingeben...
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
#Für Reproduzierbarkeit
np.random.seed(42)
#a) Daten generieren
#Generiere Tagesdaten für das Schaltjahr 2024 (366 Tage)
dates = pd.date_range(start='2024-01-01', end='2024-12-31', freq='D')
#Zufallswerte aus N(18,2^2)
temperatures = np.random.normal(loc=18, scale=2, size=len(dates))
#Erstelle DataFrame und runde die Temperaturwerte
df = pd.DataFrame({
    'Datum': dates, 
    'Tagestemperatur': np.round(temperatures, 2)
})
#Speichere die Daten in der Datei "daten-p3-aus-1.txt" mit Semikolon als Trenner
df.to_csv('daten-p3-aus-1.txt', sep=';', index=False)
#b) Daten einlesen und visualisieren
#Lese die Datei mit Angabe von parse_dates für die "Datum"-Spalte
df = pd.read_csv('daten-p3-aus-1.txt', sep=';', parse_dates=['Datum'])
#Ausgabe der wichtigsten statistischen Kennzahlen
print(df.describe())
#Scatterplot der Daten mit Seaborn
plt.figure(figsize=(14, 8))
sns.scatterplot(x='Datum', y='Tagestemperatur', data=df, label='Temperaturwerte')
plt.title("Scatterplot der Tagestemperaturen (2024)")
plt.xlabel("Datum")
plt.ylabel("Tagestemperatur (°C)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
#Boxplot der Temperaturwerte
plt.figure(figsize=(6, 8))
sns.boxplot(y='Tagestemperatur', data=df, width=0.3)
#Berechne und visualisiere das arithmetische Mittel
mean_val = df['Tagestemperatur'].mean()
plt.scatter(0, mean_val, color='red', s=100, zorder=10, label=f'Mean: {mean_val:.2f}')
plt.legend()
plt.title("Boxplot der Tagestemperaturen (2024)")
plt.ylabel("Tagestemperatur (°C)")
plt.show()
```
