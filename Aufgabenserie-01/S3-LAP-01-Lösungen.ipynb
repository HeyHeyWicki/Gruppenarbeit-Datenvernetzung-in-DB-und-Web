## Aufgabe 3
Erstelle dein erstes Pythonprogramm, welches dich mit den Worten: "Hallo, dies ist mein erstes Pythonprogramm!" begrüsst respektive auf dem Bildschirm ausgibt.
#### a) Variante 1 ohne Variable und ohne Eingabeaufforderung

```python
# wir geben mit dem Befehl 'print' und dem Parameter (hier ein String/Zeichenkette)
# die Begrüssung auf dem Bildschirm aus.
print('Hallo, dies ist mein erstes Pythonprogramm!')
```

#### b) Variante 2 mit einer Variablen ohne Eingabeaufforderung

```python
# wir weisen der Variablen 'begruessung' die Zeichenkette mit dem
# gewünschten Worten zu
begruessung = 'Hallo, dies ist mein erstes Pythonprogramm!'

# wir geben mit dem Befehl 'print' und dem Parameter (hier 'begruessung')
# die geforderten Worte auf dem Bildschirm aus.
print(begruessung)
```

#### c) Variante 3 mit einer Variablen und Eingabeaufforderung

```python
# wir geben den Text "Gib den Begrüssungstext ein:" auf dem Bildschirm aus und verlangen
# vom Benutzer die Eingabe ('input'-Befehl) des gwünschten Begrüssungstextes, welcher 
# mit Drücken auf die Enter-Taste abgeschlossen wird.
begruessung = input('Gib den Begrüssungstext ein:')

# wir geben mit dem Befehl 'print' und dem Parameter (hier 'begruessung')
# die geforderten Worte auf dem Bildschirm aus.
print(begruessung)
```

## Aufgabe 4
Erstelle ein Pythonprogramm, welches dich nach deinem Vornamen fragt und anschliessend folgendes ausgibt: Zuerst soll der Text 'Hallo' erscheinen, dann dein Vorname.


```python
# wir weisen der Variablen 'begruessung' die Zeichenkette 'Hallo' zu.
begruessung = 'Hallo'

# jetzt fragen wir nach deinem Vornamen
vorname = input('Wie lautet dein Vorname?')

# und jetzt geben wir die beiden Zeichenketten aus
print(begruessung, vorname)
```
