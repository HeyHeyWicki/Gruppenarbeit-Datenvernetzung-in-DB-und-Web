## Aufgabe 1 - Fehlerbehandlung (Laufzeitfehler)
### a) Eigene Eingabe
Das folgende Programm soll so erweitert werden, dass ganze Zahlen als Zähler und Nenner eingegeben werden können.

```python
from fractions import Fraction

print(Fraction(5, 6) + Fraction(1, 3))      # 5/6 + 1/3
print(Fraction(1, 3) - Fraction(2, 3))      # 1/3 - 2/3
print(Fraction(3, 4) * Fraction(2, 3))      # 3/4 * 2/3
print(Fraction(2, 7) / Fraction(2, 9))      # 2/7 / 2/9
```

```python
from fractions import Fraction

# Eingabe des ersten Bruchs
zaehler1 = int(input("Gib den Zähler des ersten Bruchs ein: "))
nenner1 = int(input("Gib den Nenner des ersten Bruchs ein: "))

# Eingabe des zweiten Bruchs
zaehler2 = int(input("Gib den Zähler des zweiten Bruchs ein: "))
nenner2 = int(input("Gib den Nenner des zweiten Bruchs ein: "))
    
# Erstellen der Brüche
bruch1 = Fraction(zaehler1, nenner1)
bruch2 = Fraction(zaehler2, nenner2)
    
# Berechnungen
print(f"{bruch1} + {bruch2} = {bruch1 + bruch2}")
print(f"{bruch1} - {bruch2} = {bruch1 - bruch2}")
print(f"{bruch1} * {bruch2} = {bruch1 * bruch2}")
print(f"{bruch1} / {bruch2} = {bruch1 / bruch2}")
```

### b) Fehlerbehandlung (Laufzeitfehler)
Jetzt soll das Programm so erweitert werden, dass durch eine Fehlerbehandlung keine unkontrollierten Fehler mehr entstehen können.

```python
from fractions import Fraction

try:
    # Eingabe des ersten Bruchs
    zaehler1 = int(input("Gib den Zähler des ersten Bruchs ein: "))
    nenner1 = int(input("Gib den Nenner des ersten Bruchs ein: "))
    
    # Eingabe des zweiten Bruchs
    zaehler2 = int(input("Gib den Zähler des zweiten Bruchs ein: "))
    nenner2 = int(input("Gib den Nenner des zweiten Bruchs ein: "))
    
    # Erstellen der Brüche
    bruch1 = Fraction(zaehler1, nenner1)
    bruch2 = Fraction(zaehler2, nenner2)
    
    # Berechnungen
    print(f"{bruch1} + {bruch2} = {bruch1 + bruch2}")
    print(f"{bruch1} - {bruch2} = {bruch1 - bruch2}")
    print(f"{bruch1} * {bruch2} = {bruch1 * bruch2}")
    print(f"{bruch1} / {bruch2} = {bruch1 / bruch2}")

except ValueError:
    print("Du hast keine gültige ganze Zahl eingegeben.")
except ZeroDivisionError as zde:
    print("Division durch 0 ist nicht definiert. Fehlermeldung:", zde)
except Exception as e:
    print(f"Ein unerwarteter Fehler ist aufgetreten: {e}")
finally:
    print("Das Programm wurde beendet.")
```

## Aufgabe 2 - Fehlerbehandlung (Laufzeitfehler)
Das folgende Programm soll so erweitert werden, dass mögliche Laufzeitfehler abgefangen werden. Dabei soll die Fallunterscheidung mit Hilfe der Diskriminante wegfallen.

```python
# für die Berechnung der Quadratwurzel benötigen wir das Modul "math"
import math as math

a = float(input("Gib die den Koeffizienten a ein:"))
b = float(input("Gib die den Koeffizienten b ein:"))
c = float(input("Gib die den Koeffizienten c ein:"))

diskriminante = b ** 2 - 4 * a * c

if diskriminante >= 0:
    if diskriminante == 0:
        loesung = - b / (2 * a)
        print("Die Doppellösung lautet:", loesung)
    else:
        loesung_1 = (- b + math.sqrt(diskriminante)) / (2 * a)
        loesung_2 = (- b - math.sqrt(diskriminante)) / (2 * a)
        print("Die beiden Lösungen lauten:")
        print("x1=", loesung_1)
        print("x2=", loesung_2)
else:
    print("Die Gleichung besitzt keine reellen Lösungen!")
```

```python
import math

try:
    a = float(input("Gib den Koeffizienten a ein: "))
    b = float(input("Gib den Koeffizienten b ein: "))
    c = float(input("Gib den Koeffizienten c ein: "))
   
    # Berechnung der Diskriminante
    diskriminante = b ** 2 - 4 * a * c
    
    # Berechnung der Lösungen, ohne die Diskriminante zu überprüfen
    loesung_1 = (-b + math.sqrt(diskriminante)) / (2 * a)
    loesung_2 = (-b - math.sqrt(diskriminante)) / (2 * a)
    
    print("Die beiden Lösungen lauten:")
    print("x1 =", loesung_1)
    print("x2 =", loesung_2)

except ValueError as ve:
    print(f"Ungültige Eingabe: {ve}")
except ZeroDivisionError:
    print("Division durch Null aufgetreten. Der Koeffizient a darf nicht 0 sein.")
except Exception as e:
    print(f"Ein unerwarteter Fehler ist aufgetreten: {e}")
finally:
    print("Das Programm wurde beendet.")
```

Ein Türsensor gibt den Status einer Tür an ("offen" oder "geschlossen"). Schreibe ein Programm, das den aktuellen Status überprüft und eine Meldung ausgibt. Das Programm soll dich nach einem möglich Status befragen und die Eingabe überprüfen, ob ein gültiger Status eingegeben wurde.

Setze in Zeile 2 einen _Haltepunkt_ und debugge diese Zelle und lass dir die entsprechenden Variableninhalte im Einzelschritt anzeigen, so dass du den Programmfluss nachvollziehen kannst. Das heisst, definiere die folgenden Ausdrücke zum überwachen:
-   gueltiger_status
-   gueltig
-   tuer_status.

Vegiss nicht, den Haltepunkt wieder zu löschen.   


```python
# Gültige Stati festlegen
gueltiger_status = ['offen', 'geschlossen']

gueltig = False
while not(gueltig):
    tuer_status = input('Gib einen Türstatus ein: "offen" / "geschlossen":')
    if tuer_status in gueltiger_status:
        gueltig = True
        
# Selektion zur Überprüfung des Türstatus
if tuer_status == 'offen':
    print('Achtung: Die Tür ist offen!')
else:
    print('Die Tür ist geschlossen.')    
```

#### a) Zufallszahlen nach Vorgabe generieren
Erstelle ein Python-Programm, mit welchem eine Anzahl (n = 1500) vorgegebener Zufallszahlen mit bestimmtem Mittelwert und Streueung (Standardabweichung) berechnet werden. Diese sollen auf eine Nachkommastelle gerundet in einer Datei _daten-aus-8-6.txt_ gespeichert werden.

```python
import numpy as np
import pandas as pd

# die Zufallsdaten der Maschinenproduktion
datei_ausgabe         = "Daten/daten-aus-8-6.txt"
# diejenigen Daten, welche die Vorgabe nicht erfüllen (zu wenig abgefüllt)
datei_ausgabe_weniger = "Daten/daten-aus-8-6-weniger.txt"
# diejenigen Daten, welche die Vorgabe nicht erfüllen (zu viel abgefüllt)
datei_ausgabe_mehr    = "Daten/daten-aus-8-6-mehr.txt"

mittelw  = 100      # der Mittelwert (Sollwert) der zufällig generierten Zahlen wird festgelegt
standabw = 3        # die Streuung wird hier auf drei festgelegt
n        = 1500     # die Anzahl der Zufallszahlen wird definiert

# bestimmt die Zufallszahlen mit den gewünschten Parametern/Verteilung in einem Array
np.random.seed(1)
daten    = np.random.normal(mittelw, standabw, n)
gerundet = np.around(daten, decimals = 1)

# die gerundeten Zufallszahlen werden in eine Textdatei gespeichert
# np.savetxt(datei_ausgabe, gerundet, fmt = '%2.1f')
gerundet = pd.DataFrame(gerundet)
gerundet.to_csv(datei_ausgabe, decimal='.', header=False, index=0)
```

#### g) Verlust oder Gewinn?
Wie würdest du vorgehen, wenn du weisst, dass 1 g Schokolade 0.80 CHF kostet, um auszurechnen, wieviel Verlust oder Gewinn die Firma mit dieser Produktionsserie ausweist? Brauchst du noch weitere Daten?

```python
import pandas as pd

datei_mehr      = "Daten/daten-aus-8-6-mehr.txt"
datei_weniger   = "Daten/daten-aus-8-6-weniger.txt"

untere_grenze =  97
obere_grenze  = 103
preis_gramm   = 0.8

# Die generierten Daten werden eingelesen
mehr    = pd.read_csv(datei_mehr, decimal ='.', sep=';')
weniger = pd.read_csv(datei_weniger, decimal ='.', sep=';')

# wir berechnen die jeweilige Summe
summe_mehr    = mehr['Masse'].sum()
summe_weniger = weniger['Masse'].sum()

# wir müssen jetzt noch je Zwischenresultat die erlaubte Masse
# je Packung subtrahieren, so dass wir die effektiven Differenzen
# vorliegen haben.
summe_mehr_eff = summe_mehr - (len(mehr) * obere_grenze)
summe_weniger_eff = (len(weniger) * untere_grenze) - summe_weniger

resultat_mehr = summe_mehr_eff * preis_gramm
resultat_weniger = summe_weniger_eff * preis_gramm

print(f"Zu viel produziert in CHF: {resultat_mehr:.2f}")
print(f"Zu wenig produziert in CHF: {resultat_weniger:.2f}", )
```

## Aufgabe 7 - Wohnungstemperaturen und Heizölverbrauch

#### a) Daten generieren
Erstelle ein Pythonprogramm, mit welchem du die Temperaturen und den Ölverbrauch (durchschnittlich pro Tag) für eine Wohnung im Gebäude nach den Vorgaben:
-   Temperatur:
    -   Mittelwert = 20 Grad Celsius
    -   Standardabweichung (Streuung) = 2 Grad Celsius
    -   Anzahl Werte n = 365 (1 Jahr)

-   Ölverbrauch:
    -   Mittelwert = 8 l
    -   Standardabweichung (Streuung) = 2 Grad Celsius
    -   Anzahl Werte n = 365 (1 Jahr)

zufällig und ohne Abhängigkeit von der Jahreszeit und unabhängig voneinander generieren kannst. Speichere die Daten in einer Datei mit dem Namen _daten_aus_8_7.txt_ ab.

```python
import numpy as np
import pandas as pd

# die Zufallsdaten der Wohnungstemperaturen und des Heizölverbrauches
datei_ausgabe = "Daten/daten-aus-8-7.txt"

mittelw_oel   =   8     # der Mittelwert (Sollwert) der zufällig generierten Zahlen wird festgelegt
standabw_oel  =   1     # die Streuung 

mittelw_temp  =  20     # der Mittelwert (Sollwert) der zufällig generierten Zahlen wird festgelegt
standabw_temp =   2     # die Streuung 
n             = 365     # die Anzahl der Zufallszahlen wird definiert

# bestimmt die Zufallszahlen mit den gewünschten Parametern/Verteilung in einem Array
np.random.seed(1)
daten_oel = np.random.normal(mittelw_oel, standabw_oel, n)
daten_temp = np.random.normal(mittelw_temp, standabw_temp, n)
# rundet auf eine Nachkommastelle
gerundet_oel = np.around(daten_oel, decimals=1)
gerundet_temp = np.around(daten_temp, decimals=1)

# Erstellen eines DataFrames mit beiden Datensätzen
gerundet = pd.DataFrame({
    'Oelverbrauch': gerundet_oel,
    'Temperatur': gerundet_temp
})

# die gerundeten Zufallszahlen werden in eine Textdatei gespeichert
gerundet.to_csv(datei_ausgabe, decimal='.', header=True, index=False)
```
