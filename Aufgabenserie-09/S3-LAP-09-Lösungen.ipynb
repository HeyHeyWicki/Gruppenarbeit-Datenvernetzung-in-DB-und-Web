## Aufgabe 1 - Daten visualisieren - 2D-Funktionsplots
Häufig sollen Daten visualisiert werden und liefern so ein besseres Bild, als wenn nur sogenannte Parameter (Lage- und Streu-Parameter) berechnet werden. ### a) beliebige Funktion darstellen
Erstelle ein Programm, mit welchem eine mathematische Funktion $f(x)=-x^2-4x+3$ grafisch dargestellt werden kann. Dazu kannst du die beiden Bibliotheken numpy und matplotlib einsetzen. Vor allem matplotlib dient der grafischen Darstellung.

```python
import numpy as np
import matplotlib.pyplot as plt

# Definitionsbereich
x = np.linspace(-10, 7, 500)

# Funktionsgleichung, nicht als Python-Funktion definiert
y = - x**2 - 4*x + 3

# wir können zum "schöneren" Darstellen der Funktion sogenannten
# LaTex-Code benutzen. Dieser wird in zwei "$" (Dollar-Zeichen)
# eingeschlossen:

# Grafik-Titel
plt.title('$f(x)=-x^2 - 4x + 3$')

# Daten visualisieren, lw = Linienbreite (line width)
plt.plot(x, y, color = "blue", lw = 2)

plt.xlabel('x')
plt.ylabel('y=f(x)', rotation=90)
plt.grid(True)
plt.show()
```

### b) Mehrere Funktionen in einer Ausgabe erstellen
Stelle die folgenden vier Funktionen in einer Grafik dar:
-   $f_1(x)=x-2$
-   $f_2(x)=x^2 -4$
-   $f_3(x)=1/x$
-   $f_4(x)=e^x$


```python
import numpy as np
import matplotlib.pyplot as plt

# Definitionsbereich
x = np.linspace(-5.5, 5.5, 500)

# Funktionsgleichungen
y1 = x - 2
y2 = x**2 - 4
y3 = 1.0 / x
y4 = np.exp(x)

# mehrere Funktionen darstellen, mehrere Unterdiagramme
plt.subplots(figsize = (8,8))

# links oben
plt.subplot(2, 2, 1)
plt.title(r'$y=x-2$')
plt.plot(x, y1, color = "blue", lw = 2)
plt.ylabel('y', rotation = 0)
plt.grid(True)

# rechts oben
plt.subplot(2, 2, 2)
plt.title(r'$y=x^{2}-4$')
plt.plot(x, y2, color = "red", lw = 2)
plt.grid(True)

#links unten
plt.subplot(2, 2, 3)
plt.title(r'$y=\frac{1}{x}$')
plt.plot(x, y3, color = "green", lw = 2)
plt.xlim(0.1,2)
plt.ylim(0.1,10)
plt.xlabel('x')
plt.ylabel('y',rotation = 0)
plt.grid(True)

#rechts unten
plt.subplot(2, 2, 4)
plt.title(r'$y=e^{x}$')
plt.plot(x, y4, color= "black", lw = 2)
plt.xlabel('x')
plt.grid(True)

# genügend Platz zwischen den Unterdiagrammen reservieren
plt.tight_layout()
plt.show()
```

### c) vereinfacht mit Qodo Gen

### d) alle vier Funktionen in einem Koordinatensystem
Und hier abschliessend ein, ebenfalls komplett durch Qodo Gen erstelltes Pythonprogramm, für die Darstellung aller Funktionen in einem Koordinatensystem.

```python
# Quelle: Qodo Gen

import numpy as np
import matplotlib.pyplot as plt

# Definitionsbereich
x = np.linspace(-5.5, 5.5, 500)

# Funktionsgleichungen
functions = [
    (x - 2, r'$y=x-2$', "blue"),
    (x**2 - 4, r'$y=x^{2}-4$', "red"),
    (1.0 / x, r'$y=\frac{1}{x}$', "green"),
    (np.exp(x), r'$y=e^{x}$', "black")
]

# Erstellen des Plots
plt.figure(figsize=(10, 6))

for y, label, color in functions:
    plt.plot(x, y, label=label, color=color, lw=2)

plt.xlim(-5.5, 5.5)
plt.ylim(-10, 10)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Darstellung mehrerer Funktionen')
plt.legend()
plt.grid(True)
plt.show()
```

### e) mit nummerierter Legende
Die Legende ist für Farbenblinde nicht optimal. Ersetze diese Farben durch eine numerische Angabe und beschrifte die Funktionen entsprechend mit dieser Zahl.

```python
# Quelle: Qodo Gen 

import numpy as np
import matplotlib.pyplot as plt

# Definitionsbereich
x = np.linspace(-5.5, 5.5, 500)

# Funktionsgleichungen
functions = [
    (x - 2, r'$1: y=x-2$', "blue"),
    (x**2 - 4, r'$2: y=x^{2}-4$', "red"),
    (1.0 / x, r'$3: y=\frac{1}{x}$', "green"),
    (np.exp(x), r'$4: y=e^{x}$', "black")
]

# Erstellen des Plots
plt.figure(figsize=(10, 6))

for i, (y, label, color) in enumerate(functions, 1):
    plt.plot(x, y, label=label, color=color, lw=2)
    # Position für die Annotation wählen
    if i == 3:
        # Spezielle Position für die Funktion y = 1/x
        x_pos = 2
        y_pos = 1 / x_pos
    else:
        x_pos = x[len(x) // 2]  # Mitte des Definitionsbereichs
        y_pos = y[len(x) // 2]  # Wert der Funktion an dieser Stelle
    plt.annotate(str(i), xy=(x_pos, y_pos), xytext=(x_pos + 0.5, y_pos),
                 textcoords='offset points', fontsize=12, color=color)

plt.xlim(-5.5, 5.5)
plt.ylim(-10, 10)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Darstellung mehrerer Funktionen')
plt.legend()
plt.grid(True)
plt.show()
```

## Aufgabe 2 - Daten einlesen, visualisieren - 2D-Funktionsplots
Lies die Daten aus der letzten Aufgabenserie _daten-aus-8-5.txt_ ein, gib die wichtigsten, beschreibenden Statistkparameter an und visualisiere diese Daten.

```python
import pandas as pd

# allenfalls musst du das Datenverzeichnis anpassen. 
datei_eingabe       = "Daten/daten-aus-8-5.txt"

# Einlesen der Daten mit Hilfe der importierten Bibliothek pandas,
# auf welche wir wegen des obigen Importbefehls mit pd zugreifen können.
daten = pd.read_csv(datei_eingabe, decimal ='.')
```

```python
daten.describe()
```

### a) eingelesene Daten (eine Variable) als Punkteplot darstellen
Auf der x-Achse soll der Index, das ist die fortlaufende Zahl von Null bis zur Anzahl der Datenwerte angezeigt werden. Korrespondieren dazu soll der Datenwert auf der y-Achse angezeigt werden.

```python
import matplotlib.pyplot as plt

# Erstellen des Linienplots
plt.figure(figsize=(10, 6))
plt.plot(daten, label="Datenverlauf")
plt.title("Punkteplot der eingelesenen Daten")
plt.xlabel("Index")
plt.ylabel("Werte")
plt.legend()
plt.grid(True)
plt.show()
```

### b ) Visualisierung ergänzen
Ergänze den Plot durch das Einzeichnen des arithmetischen Mittelwertes und des absoluten Minimums und Maximums der Daten.

```python
import matplotlib.pyplot as plt

# Erstellen des Linienplots
plt.figure(figsize=(10, 6))
plt.plot(daten, label="Datenverlauf", color='b')

mean = daten.mean().values[0]
min_val = daten.min().values[0]
max_val = daten.max().values[0]

plt.axhline(mean, color='r', linestyle='--', label='Mittelwert: {:.2f}'.format(mean))
plt.axhline(min_val, color='c', linestyle=':', label='Min: {:.2f}'.format(min_val))
plt.axhline(max_val, color='c', linestyle=':', label='Max: {:.2f}'.format(max_val))

plt.title("Visulaisierung der Daten mit statistischen Parametern")
plt.xlabel("Index")
plt.ylabel("Werte")
plt.legend()
plt.grid(True)

plt.show()
```

### c) Histogramm
Damit wir die Datenhäufigkeiten besser erkennen können, zeichnen wir ein sogenanntes [Histogramm](https://de.wikipedia.org/wiki/Histogramm) ein. Wir erinnern uns, die Daten haben wir selber als normalverteilte Zufallszahlen generiert. In der Grafik erkennen wir deshalb _fast_ eine Gauss'sche Glockenkurve, welche für eine Normalverteilung steht.

```python
import matplotlib.pyplot as plt

# erstellen eines Histogramms
plt.figure(figsize=(10, 6))
plt.hist(daten, bins=30, color='lightgrey', edgecolor='black', label='Histogramm')

plt.title("Visulaisierung der Daten mit Hilfe eines Histogramms")
plt.xlabel("Index")
plt.ylabel("Werte")
plt.legend()
plt.grid(True)
plt.show()
```

### d) Normalverteilung
Damit wir unsere generierten Daten mit der Gauss'schen Normalverteilung vergleichen können, müssen wir diese zuerst berechnen und im gleichen Koordinatensystem darstellen.

```python
# für die Darstellung der Normalverteilung, benötigen wir die Funktion norm aus der Bibliothek scipy.stats
from scipy.stats import norm

# Berechnung der statistischen Parameter
mean = daten.mean()
standard = daten.std()

# Erstellen des Histogramms
plt.figure(figsize=(10, 6))
plt.hist(daten, bins=30, color='lightgrey', edgecolor='black', density=True, label='Histogramm')

# Normalverteilung berechnen und plotten
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mean, standard)
plt.plot(x, p, 'r', linewidth=2, label='Normalverteilung')

plt.title("Visualisierung der Daten mit Histogramm und Normalverteilung")
plt.xlabel("Werte")
plt.ylabel("Dichte")
plt.legend()
plt.grid(True)
plt.show()
```

### e) Berechnung der Wahrscheinlichkeit (theoretisch)
Wir wollen nun noch berechnen, wie gross die Wahrscheinlichkeit ist, dass unser Wert in einem vorgegebenen Intervall liegt. Dies können wir ebenfalls mit Hilfe der Funktion `norm.cdf()` ([kumulative Verteilungsfunktion](https://de.wikipedia.org/wiki/Verteilungsfunktion)) aus `scipy.stats` erledigen.

```python
from scipy.stats import norm

# Berechnung der statistischen Parameter
mean = daten.mean()
standard = daten.std()

# Intervallgrenzen
a = 22.5
b = 27.5

# Berechnung der Wahrscheinlichkeit, dass die Zufallsvariable im Intervall [a, b] liegt
wahrscheinlichkeit = norm.cdf(b, mean, standard) - norm.cdf(a, mean, standard)

print(f"Die Wahrscheinlichkeit, dass die Zufallsvariable im Intervall [{a}, {b}] liegt, beträgt {wahrscheinlichkeit}.")
```

## Aufgabe 4 - Daten einlesen, auswerten und vergleichen (Würfelstichproben)
In den beiliegenden Dateien haben fünf Personen mit einem Würfel (Augenzahlen 1 bis 6) gewürfelt. In jeder Datenzeile sind vier Wurfergebnisse aufgelistet. Die jeweiligen Ergebnisse sind in diesen Dateien gespeichert: 

-   «daten-ein-wuerfel-1.txt»
-   «daten-ein-wuerfel-2.txt»
-   «daten-ein-wuerfel-3.txt»
-   «daten-ein-wuerfel-4.txt»
-   «daten-ein-wuerfel-5.txt».

1.  Untersuche diese Daten und kategorisiere sie.
2.  Um welche Verteilungen handelt es sich bei den einzelnen Kategorien?
3.  Visualisiere deine Resultate und stelle die Häufigkeiten als Säulendiagramme respektive als Histogramme dar.
4.  Welche Würfel waren «gezinkt»?

### a) Daten einlesen und vorbereiten
Lies die obenstehenden Dateien ein und bereite die Berechnungen vor. Es gibt in jeder Datei sechs Ereignisse, welche in den folgenden Spalten aufgelistet sind:
-   EINS
-   ZWEI
-   DREI
-   VIER
-   FUENF
-   SECHS

Die erste Spalte __Runde__ steht für die Anzahl Würfelrunden, in welcher jeweils vier Mal ein Würfel geworfen wird. Diese soll nicht importiert werden, das heisst, wir setzen beim Einlesen den Parameter `index_col=0`.

```python
import matplotlib.pyplot as plt
import pandas as pd

# Daten aus den einzelnen Dateien einlesen
eingabe_1 = pd.read_csv('Daten/daten-ein-wuerfel-1.txt', delimiter=';', index_col = 0)
eingabe_2 = pd.read_csv('Daten/daten-ein-wuerfel-2.txt', delimiter=';', index_col = 0)
eingabe_3 = pd.read_csv('Daten/daten-ein-wuerfel-3.txt', delimiter=';', index_col = 0)
eingabe_4 = pd.read_csv('Daten/daten-ein-wuerfel-4.txt', delimiter=';', index_col = 0)
eingabe_5 = pd.read_csv('Daten/daten-ein-wuerfel-5.txt', delimiter=';', index_col = 0)

# wir ermitteln die Spaltennamen
spalten = list(eingabe_1.columns)

# Funktion zur Berechnung der Summe
def berechne_Summe(eingabe):
    summe = []
    for i in spalten:
        s = eingabe[i].sum()
        summe.append(s)
    return summe
```

### b) Summenhäufigkeiten berechnen

```python
# Berechnung der Summen
summe_1   = berechne_Summe(eingabe_1)
summe_2   = berechne_Summe(eingabe_2)
summe_3   = berechne_Summe(eingabe_3)
summe_4   = berechne_Summe(eingabe_4)
summe_5   = berechne_Summe(eingabe_5)

# Erstellen der Dictionarys
haeufig_1 = dict(zip(spalten, summe_1))
haeufig_2 = dict(zip(spalten, summe_2))
haeufig_3 = dict(zip(spalten, summe_3))
haeufig_4 = dict(zip(spalten, summe_4))
haeufig_5 = dict(zip(spalten, summe_5))
```

### c) Visualisierung der Daten als Säulendiagramme

```python
# Visualisierungen der einzelnen Datensätze
plt.bar(haeufig_1.keys(), haeufig_1.values(), label='Daten-1')
plt.xticks(spalten)
plt.xlabel('Ereignisse')
plt.ylabel('Anzahl')
plt.title('Anzahl der gewürfelten Auganzahlen')
plt.legend()
plt.show()

plt.bar(haeufig_2.keys(), haeufig_2.values(), label='Daten-2')
plt.xlabel('Ereignisse')
plt.ylabel('Anzahl')
plt.title('Anzahl der gewürfelten Auganzahlen')
plt.legend()
plt.show()

plt.bar(haeufig_3.keys(), haeufig_3.values(), label='Daten-3')
plt.xlabel('Ereignisse')
plt.ylabel('Anzahl')
plt.title('Anzahl der gewürfelten Auganzahlen')
plt.legend()
plt.show()

plt.bar(haeufig_4.keys(), haeufig_4.values(), label='Daten-4')
plt.xlabel('Ereignisse')
plt.ylabel('Anzahl')
plt.title('Anzahl der gewürfelten Auganzahlen')
plt.legend()
plt.show()

plt.bar(haeufig_5.keys(), haeufig_5.values(), label='Daten-5')
plt.xlabel('Ereignisse')
plt.ylabel('Anzahl')
plt.title('Anzahl der gewürfelten Auganzahlen')
plt.legend()
plt.show()
```

### d) Ausgabe der verschiedenen Verteilungen der Daten

Es gibt fünf verschiedene Kategorien:
-   Daten-1: annähernde Gleichverteilung
-   Daten-2: rechtsschiefe Verteilung
-   Daten-3: Normalverteilung, symmetrisch
-   Daten-4: linksschiefe Verteilung
-   Daten-5: symmetrische Verteilung
