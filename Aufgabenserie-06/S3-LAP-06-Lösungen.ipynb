## Aufgabe 1 - Verwaltung von Sensoren (Smart-Building)
Ein Gebäude hat n Räume, und jeder Raum hat Sensoren, die nacheinander überprüft werden müssen. Prüfe rekursiv alle Sensoren in den Räumen und gib aus, welcher Raum gerade bearbeitet wird.

```python
raeume = int(input("Gib die Anzahl Räume ein:"))
def sensoren_pruefen(raum):
    # Abbruchbedingung
    if raum == 0:
        return
    # Funktion ruft sich selbst auf
    sensoren_pruefen(raum - 1)
    print(f"Sensoren in Raum {raum} werden geprüft.")

# Aufruf der Funktion
sensoren_pruefen(raeume)
```

## Aufgabe 2 - Wartungsplanung für Räume
Ein Gebäude hat n Räume, und die Reihenfolge der Wartung dieser Räume muss so geplant werden, dass jede mögliche Reihenfolge betrachtet werden kann. Die Anzahl der möglichen Reihenfolgen lässt sich durch die Fakultät berechnen.

#### Variante 1 - iterativ

```python
zahl = int(input("Berechne die Fakultät von: "))

# das Produkt wird initialisiert
fak = 1

# wichtig: obere Grenze läuft standargemäss nur bis zahl-1
for index in range(1, zahl + 1):
    # das Produkt wird faktorweise erweitert    
    fak = fak * index

# Ausgabe der Fakultät              
print("Die Fakulät von ", zahl, "ist:", fak)   
```

#### Variante 2 - rekursiv

```python
zahl = int(input("Berechne die Fakultät von: "))
def fakultaet(n):
    # wir stellen hier sicher, dass wir keine Endlosschleife
    # generieren!
    if n <= 1:
        return 1
    else:
        # hier rufen wir die Funktion selbst wieder auf
        return n * fakultaet(n-1)

print("Die Fakulät von ", zahl, "ist:", fakultaet(zahl))
```

#### Anwendung in einem Gebäude

-   Optimierung der Wartungsreihenfolge: Das Ergebnis zeigt, wie viele mögliche Reihenfolgen betrachtet werden müssen, um die effizienteste Wartung zu planen.
-   Analyse von Szenarien: Die Fakultät hilft, alle Permutationen durchzuspielen, z. B. bei der Sensorinstallation oder bei der Anordnung von Steuergeräten in einem Gebäude.

Erläuterungen:
-   Fakultät-Funktion: Berechnet die Anzahl der Permutationen (mögliche Reihenfolgen).
-   _itertools.permutations_: Erzeugt alle möglichen Wartungsreihenfolgen der Räume als Tupel. Verwende die Funktion _permutations_ aus der Bibliothek [itertools](https://docs.python.org/3/library/itertools.html).

```python
from itertools import permutations

def fakultaet(n):
    if n == 0 or n == 1:
        return 1
    return n * fakultaet(n - 1)

def wartungsreihenfolgen(raeume):
    # Berechnung der Fakultät (Anzahl der möglichen Reihenfolgen)
    moeglichkeiten = fakultaet(len(raeume))
    print(f"Es gibt {moeglichkeiten} mögliche Reihenfolgen für die Wartung von {len(raeume)} Räumen.")
    
    # Erzeugung aller möglichen Wartungsreihenfolgen
    alle_reihenfolgen = list(permutations(raeume))
    print("\nMögliche Wartungsreihenfolgen:")
    for reihenfolge in alle_reihenfolgen:
        print(reihenfolge)

# Beispiel: Wartung von 3 Räumen
raeume = ["Eiger (R1)", "Mönch (R2)", "Jungfrau (R3)"]
wartungsreihenfolgen(raeume)
```

#### Ohne Funktion

```python
# ohne eigene Funktion
# Eingabe
zahl = int(input("Nach welcher Zahl zwischen 1 und 10 soll gesucht werden:"))

# in der folgenden Liste sollen Zahlen gesucht werden
zahlen = [9, 2, 7, 4, 6, 5, 3, 8, 1, 10]

anzahl   = len(zahlen)
gefunden = False

for i in range(anzahl):
    if zahlen[i] == zahl:
        gefunden = True
        # wenn wir das gesuchte Element/Zahl gefunden haben,
        # können wir die Schleife verlassen; dies geschieht durch
        # den Befehl "break". 
        break

if gefunden:
    print("Die Zahl", zahl, "ist in der Liste und wurde gefunden")
else:
    print("Die Zahl", zahl, "ist nicht in der Liste und konnte somit nicht gefunden werden!")
```

#### Mit Funktion

```python
# mit eigender Funktion
# Eingabe
zahl = int(input("Nach welcher Zahl zwischen 1 und 10 soll gesucht werden:"))

# in der folhenden Liste sollen Zahlen gesucht werden
zahlen = [9, 2, 7, 4, 6, 5, 3, 8, 1, 10]

def lineare_suche(element, zahlen_liste):
	# Anzahl Elemente
	anzahl = len(zahlen_liste)
	# Boole'sche Variable für die Übereinstimmung
	gefunden = False				
	# jetzt vergleichen wir schrittweise jedes Element in der
	# Liste (Liste) mit dem zu suchenden (Zahl)
	for i in range(anzahl):
		# vergleiche das aktuelle mit dem gesuchten Element
		if zahlen_liste[i] == element:
			# Übereinstimmung, wir setzen unsere Hilfsvariable
			# auf True, das Element kennen wir ja
			gefunden = True
	# Wir geben den Wert der Variablen "gefunden" zurück
	return gefunden

if lineare_suche(zahl, zahlen) == True:
	print("Die Zahl", zahl, "ist in der Liste und wurde gefunden")
else:
	print("Die Zahl", zahl, "ist nicht in der Liste und konnte somit nicht gefunden werden!")
```

## Aufgabe 4 - Binäre Suche in Listen
Versuche dir zuerst ein Bild von der binären Suche zu machen. Vielleicht gelingt es an 
folgendem Beispiel: Du denkst dir eine Zahl zwichen 1 und 100 aus. Jemand anderes soll nun 
diese Zahl erraten. Du gibst jeweils nur die Antworten, ob die geratene Zahl grösser, kleiner 
oder gleich ist. Wie könnte man hier am schnellsten zur Geheimzahl gelangen?<br>

<span style="color: green;">
a)
</span>
Erstelle ein Programm, mit welchem in einer Liste von 10 Zahlen zwischen 1 (inklusive) und 10 (inklusive) nach einer einzugebenden Zahl gesucht werden kann. Falls die Zahl in der Auswahl vorkommt, soll das Programm dies bestätigen. Du kannst die Lösung mit oder ohne eigene Funktion lösen aber jetzt mit einer binären Suche und iterativ.

#### Iterative Lösung mit Funktion

```python
# Eingabe
zahl = int(input("Nach welcher Zahl zwischen 1 und 10 soll gesucht werden:"))

# in der folgenden Liste sollen Zahlen gesucht werden
zahlen = [9, 2, 7, 4, 6, 5, 3, 8, 1, 10]

anzahl = len(zahlen)

# iterative Version mit Funktion
def binaere_suche(zahl, liste):

	# Startpositionen
	links  = 0
	rechts = len(liste) - 1

	# wir sortieren die Elemente der Liste in
	# aufsteigender Reihenfolge (Standard) und
	# speichern diese in einer neuen Variablen,
	# falls nicht, würden wir die ursprüngliche Liste über-
	# schreiben.
	sort_liste = sorted(liste)

	# Initialisierung der Laufvariablen
	i = 0
	gefunden = False
	## Wiederhole folgende Schritte
	while i <= rechts:
		## mittlere Position bestimmen
		mitte = (links + rechts) // 2
		if sort_liste[mitte] == zahl:
			## wir haben die Zahl gefunden und können die
			# Schleife verlassen
			gefunden = True
			break
		elif sort_liste[mitte] < zahl:
			## neue linke Position, legt das untere Ende des 
			# "halbierten" Intervalls fest
			links = mitte + 1
		else:
			## neue rechte Position, legt das obere Ende des
			# "halbierten" Intervalls fest 
			rechts = mitte - 1
		# erhöhe die Laufvariable um 1
		i = i + 1
	# die Zahl wurde nicht gefunden	
	return gefunden

resultat = binaere_suche(zahl, zahlen)

if resultat:
	 print("Die Zahl", zahl, "ist in der Liste und wurde gefunden")
else:
	 print("Die Zahl", zahl, "ist nicht in der Liste und konnte somit nicht gefunden werden!")
	
```

#### Rekursive Lösung

<span style="color: green;">
b)
</span>
Erstelle ein Programm, mit welchem in einer Liste von 10 Zahlen zwischen 1 (inklusive) und 10 (inklusive) nach einer einzugebenden Zahl gesucht werden kann. Falls die Zahl in der Auswahl vorkommt, soll das Programm dies bestätigen. Du kannst die Lösung mit oder ohne eigene Funktion lösen aber jetzt mit einer binären Suche und rekursiv.

```python
# Eingabe
zahl = int(input("Nach welcher Zahl zwischen 1 und 10 soll gesucht werden:"))

# in der folgenden Liste sollen Zahlen gesucht werden
zahlen = [9, 2, 7, 4, 6, 5, 3, 8, 1, 10]

# Initialisierung
links       = 0
rechts      = len(zahlen) - 1
# wir sortieren die Zahlen und speichern sie in einer anderen Variablen (zahlen_sort) ab,
# so dass wir die ursprünglichen Werte (zahlen) noch zur Verfügung haben!
zahlen_sort = sorted(zahlen)

# rekursive Version
def bin_suche_rek(zahl, links, rechts, liste):
    
    # die Suche ist beendet, da die linke Grenze grösser ist als die rechte. 
    # Wir haben das Element zahl nicht gefunden. Resultat-Wert "gefunden" zurückgeben.
    if links > rechts:
        return -1
	# wir definieren den mittleren Wert und vergleichen
    # mit dem Suchwert; ganzzahlige Division
    mitte = (links + rechts) // 2
	# Wir haben eine Übereinstimmung (Treffer), deshalb wird der Wert des 
    # mittleren Elementes zurückgegeben. Genauer: das Element mit dem mittleren Index
    if zahl == liste[mitte]:
        return mitte

    elif zahl < liste[mitte]:
        # wir rufen die Funktion mit angepassten Parameterwerten selbst auf: der rechte 
        # Teil fällt weg. Neu: rechts = mitte - 1 als neue rechte obere Grenze ersetzt.
        # REKURSIONS-AUFRUF
        return bin_suche_rek(zahl, links, mitte - 1, liste)
        
    else:
        # wir rufen die Funktion mit angepassten Parameterwerten selbst auf: der linke 
        # Teil fällt weg. Neu: links = mitte + 1 als neue linke untere Grenze ersetzt.
        # REKURSIONS-AUFRUF
        return bin_suche_rek(zahl, mitte + 1, rechts, liste)    

index = bin_suche_rek(zahl, links, rechts, zahlen_sort)

if index != -1:
    # ursprüngliche Liste zur Bestimmung der Position heranziehen; dies wäre eigentlich
    # auch gerade die kürzeste Version der Suche. Überprüfung, ob die Zahl überhaupt in der
    # Liste vorkommt, sonst gibt es eine Fehlermeldung!
    ind_eff = zahlen.index(zahl)
    print("Die Zahl", zahl, "ist in der Liste und wurde an der Position: ", ind_eff, "gefunden")
else:
    print("Die Zahl", zahl, "ist nicht in der Liste und konnte somit nicht gefunden werden!")
    
```
