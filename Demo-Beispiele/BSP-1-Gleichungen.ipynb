## Beispiel: Gleichungen dritten und höheren Grades
Gleichungen <span style="color:red">**3. Grades**</span> oder <span style="color:red">**kubische Gleichungen**</span> haben die **allgemeine Form**:

$$ 
\begin{equation}
a_3x^3+a_2x^2+a_1x+a_0=0
\end{equation}
$$

oder die **Produktdarstellung**:

$$
a_3(x-x_1)(x-x_2)(x-x_3)=0,
$$

wobei $x_1$, $x_2$ und $x_3$ die <b>Lösungen</b> der Gleichung sind. In Sonderfällen, wenn z.B. in der allgemeinen Form das **absolute Glied** $a_0=0$ ist oder die Produktdarstellung vorliegt, ist das Lösen einfach.

### <span style="color:blue">Beispiel 2.24</span>
Die Gleichung $12(x+3)(x^2-4)=0$ ist zu lösen. Gemäss der obigen **Produktdarstellung** ist sofort eine Lösung $x_1=-3$ ablesbar. Die anderen Lösungen ergeben sich aus $x^2-4=0$, das heisst:
$$
\begin{matrix}
        x^2-4   & =     & 0    & | & +4        \\
        x^2     & =     & 4    & | & \sqrt{}   \\
        x_{2,3} & =     & \pm{2}&            \\
\end{matrix}
$$

### <span style="color:blue">Beispiel 2.25</span>
Die Gleichung $3x^3-8x^2-13x=0$ ist zu lösen. Da in jedem Term $x$ als Faktor auftritt, können wir $x$ **ausklammern**. Das liefert uns die folgende Darstellung:
$$
\begin{matrix}
        x(3x^2-8x-13)   & = 0
\end{matrix}
$$

Da wir jetzt, wie bereits im vorherigen Beispiel eine _Produktdarstellung_ haben, können wir daraus folgern, dass mindestens ein Faktor Null sein muss. Das heisst, wir können direkt eine Lösung $x_1=0$ ablesen. Wollen wir nun noch die beiden anderen Lösungen berechnen, sofern sie existieren, dann haben wir die quadratische Gleichung $3x^2-8x-13=0$ zu lösen. Diese kann mit der untenstehenden Lösungsformel nach $x$ aufgelöst werden:
$$ 
        x_{2,3} = \cfrac{-b\pm{\sqrt{b^2-4ac}}}{2a}
$$
Wobei in unserem Beispiel $a=3$, $b=-8$ und $c=-13$ die Koeffizienten des Polynoms $3x^2-8x-13$ sind. 

Da die _Diskriminante_ $D=b^2-4ac$ positiv ist, gibt es zwei reelle Lösungen $x_2\approx 3.805$ und $x_3\approx -1.139$.

<b>Anmerkung</b>: Diese Folgerung, dass bei einem Produkt, welches gleich Null gesetzt wird, mindestens einer der Faktoren den Wert 0 annehmen muss, gilt nicht immer. Es gibt Operationen, bei welchen diese Folgerung nicht richtig ist. Als Beispiel soll hier die Matrizenmultiplikation angegeben werden. Es ist möglich, dass das Produkt zweier Matrizen Null ergibt, aber keine der beiden Matrizen verschwindet, also gleich 0 ist.


### Interaktive Berechnung

Nun wollen wir diese Lösungen mit Hilfe von [Python](https://www.python.org) berechnen. In dieser ersten Variante wollen wir das Ergebnis exakt berechnet haben. Das heisst, eventuelle Wurzelwerte solle als solche angegeben werden. Deshalb importieren wir aus dem Paket [sympy](https://www.sympy.org/en/index.html) die Funktion _solve_. Wir müssen [Python](https://www.python.org) noch mitteilen, dass $x$ die Variable ist, nach der aufgelöst werden soll.

#### Exakte, symbolische Berechnung

```python
from sympy.solvers import solve
from sympy import Symbol

x = Symbol('x')

# Mit dem Befehl solve lösen wir die entsprechende quadratische Gleichung
loes_1 = solve(3*x**2 - 8*x-13, x)

print("Lösungen: ", loes_1)

# Jetzt runden wir noch, damit wir die Ausgabe schneller vergleichen können
print("Lösung 1: ", round(loes_1[1],3))
print("Lösung 2: ", round(loes_1[0],3))
```

#### Numerische Berechnung

In der zweiten Variante spielt die exakte Darstellung keine Rolle. Wir wollen einfach die beiden Lösungen rechnerisch ermitteln. Hier kann uns die Bibliothek [numpy](https://numpy.org) helfen. Wir gehen wie folgt vor:

```python
import numpy as np

# Das zu lösende Polynom hat die Koeffizienten: p[0] = 3; p[1] = -8 und p[2] = -13.
koeffizienten = [3, -8, -13]
loes_2 = np.roots(koeffizienten)

print("Lösungen: ", loes_2)
```

Wenn wir jetzt noch auf drei Nachkommastellen runden, dann erhalten wir dieselben Resultate, welche wir auch oben bereits von Hand und mit [Python](https://www.python.org) exakt berechnet haben.

```python
# Die beiden Lösungswerte sind in einem Array gespeichert.
loes_2 = np.roots(koeffizienten)
# Jetzt können wir die Werte einzeln auf drei Nachkommastellen runden:
print("Lösung 1: ", round(loes_2[0],3))
# und die zweite Lösung analog. Beachte, dass der Index des Array's bei 0 beginnt!
print("Lösung 2: ", round(loes_2[1],3))
```

### Visualisierung

Die berechneten Lösungen der Gleichung $f(x)=x(3x^2-8x-13)=0$ können grafisch dargestellt werden. Diese entsprechen den $x$-Werten der Schnittpunkte der kubischen Funktion $f(x)=x(3x^2-8x-13)$ mit der x-Achse. 

Damit wir Funktionen darstellen können, benötigen wir die Bibliothek [matplotlib](https://matplotlib.org). Zudem benötigen wir wiederum die Bibliothek [numpy](https://numpy.org) und daraus den Befehl _linspace_, um die Datenwerte zur Verfügung zu stellen. Diese ist bereits oben importiert worden, weshalb wir es hier nicht mehr durchführen müssen.

```python
import matplotlib.pyplot as plt

# Bestimme die x- und y-Werte
x = np.linspace(-2,4.5,200)

# Die kubische Funktion wird definiert
y = x * (3 * x**2 - 8*x -13)

plt.title('Lösungen der Gleichung / Schnittpunkte der Funktion')
# Die Funktion wir in blauer Farbe dargestellt
plt.plot(x, y, color = 'blue')

# Die triviale Nullstelle bei x = 0 wird rot markiert
plt.plot(0, 0, "ro")
# Markiere die beiden anderen Nullstellen in roter Farbe; die beiden Lösungen haben wir ja bereits berechnet 
# und in der Variablen loes_2 gespeichert.
plt.plot(loes_2[1], 0, "ro")
plt.plot(loes_2[0], 0, "ro")

# Visualisierung
plt.plot()
plt.xlabel('x')
plt.ylabel('y=f(x)')
plt.grid(True)
plt.show()
```
