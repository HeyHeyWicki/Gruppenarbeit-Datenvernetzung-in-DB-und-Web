Du hast als Projektleiter die Aufgabe erhalten, aus den Daten des letzten Jahres eine Budget-Prognose für das nächste Jahr zu berechnen. Dabei soll der Budgetvorgabewert von Mio CHF 1'300 erreicht werden. Die saisonalen Schwankungen sollen ebenso berücksichtigt werden.
    a) Wie kannst du eine monatliche Prognose erstellen?
    b) Was ist dabei zu beachten?


```python
# Gegebene IST-Umsätze in Mio CHF
ist_umsatz = {
    "Januar": 86.5,
    "Februar": 107.1,
    "März": 94.9,
    "April": 87.8,
    "Mai": 97.4,
    "Juni": 94.2,
    "Juli": 78.2,
    "August": 110.9,
    "September": 94.2,
    "Oktober": 121.2,
    "November": 116.0,
    "Dezember": 110.0
}
# Zielbudget für das nächste Jahr in Mio CHF
zielbudget = 1300
# Berechnung des Gesamtumsatzes des letzten Jahres
gesamtumsatz = sum(ist_umsatz.values())
# Berechnung des Skalierungsfaktors
skalierungsfaktor = zielbudget / gesamtumsatz
# Berechnung der monatlichen Prognose
prognose = {monat: umsatz * skalierungsfaktor for monat, umsatz in ist_umsatz.items()}
# Ausgabe der Ergebnisse
print("Monatliche Prognose für das nächste Jahr in Mio CHF:")
for monat, wert in prognose.items():
    print(f"{monat}: {wert:.2f}")
# Gesamtprognose überprüfen
gesamtprognose = sum(prognose.values())
print(f"\nGesamtprognose für das nächste Jahr: {gesamtprognose:.2f} Mio CHF")
```

Dieses Programm führt die folgenden Schritte aus:
1. Definiert die IST-Umsätze des letzten Jahres.
2. Setzt das Zielbudget für das nächste Jahr.
3. Berechnet den Gesamtumsatz des letzten Jahres.
4. Berechnet den Skalierungsfaktor, um das Zielbudget zu erreichen.
5. Skaliert die IST-Umsätze mit dem Skalierungsfaktor, um die monatliche Prognose zu erstellen.
6. Gibt die monatliche Prognose und die Gesamtprognose aus.
Wenn du dieses Programm ausführst, erhältst du die monatliche Prognose für das nächste Jahr, die das Zielbudget von 1'300 Mio CHF berücksichtigt.
