## Aufgabe 1
Eine Gebäudeautomationsanlage speichert die aktuellen Temperaturen in den verschiedenen Räumen eines Gebäudes in einer Liste. Die Temperaturen in den Räumen sind die folgenden: 22.5; 21.0; 23.1; 19.5; 24.0; 20.3.

### a) Maximum und Minimum
Schreibe ein Programm, das die Temperaturen in der Liste ausgibt und die höchste und niedrigste Temperatur in den Räumen anzeigt.

```python
# Temperaturen in den Räumen
raum_temperaturen = [22.5, 21.0, 23.1, 19.5, 24.0, 20.3]

# Ausgabe der Temperaturen
print("Raumtemperaturen:", raum_temperaturen)

# Höchste und niedrigste Temperatur berechnen
hoechste_temp = max(raum_temperaturen)
niedrigste_temp = min(raum_temperaturen)

# Ergebnisse ausgeben
print("Höchste Temperatur:", hoechste_temp, "°C")
print("Niedrigste Temperatur:", niedrigste_temp, "°C")
```

```python
# Sollwert für die Temperatur abfragen
soll_wert = float(input('Gib einen SOLL-Temperaturwert ein:'))

# Temperaturen in den Räumen
raum_temperaturen = [22.5, 21.0, 23.1, 19.5, 24.0, 20.3]

# Höchste und niedrigste Temperatur berechnen
hoechste_temp = max(raum_temperaturen)
niedrigste_temp = min(raum_temperaturen)

# Ergebnisse ausgeben
print("Differenz zu höchstem Temperaturwert:", soll_wert, '-', hoechste_temp, '=', soll_wert - hoechste_temp, "°C")
print("Differenz zu niedrigstem Temperaturwert:", soll_wert, '-', niedrigste_temp, '=', soll_wert - niedrigste_temp, "°C")
```

## Aufgabe 4
Ein Gebäudeautomationssystem speichert die Energieverbrauchswerte verschiedener Räume in einem Dictionary, wobei der Raumname der Schlüssel und der Verbrauchswert der Wert ist. Die einzelnen Räume sind die folgenden:
+   Wohnzimmer: 12.5 kWh
+   Küche: 7.8 kWh
+   Schlafzimmer: 5.2 kWh
+   Badezimmer: 3.6 kWh


### a) Energieverbrauchswerte je Raum ausgeben
Schreibe ein Programm, das den Energieverbrauch für jeden Raum anzeigt und den Gesamtenergieverbrauch berechnet.

```python
# Energieverbrauchswerte der Räume in kWh
energieverbrauch = {
    "Wohnzimmer": 12.5,
    "Küche": 7.8,
    "Schlafzimmer": 5.2,
    "Badezimmer": 3.6
}

# Energieverbrauch für jeden Raum direkt ausgeben
print("Wohnzimmer:", energieverbrauch["Wohnzimmer"], "kWh")
print("Küche:", energieverbrauch["Küche"], "kWh")
print("Schlafzimmer:", energieverbrauch["Schlafzimmer"], "kWh")
print("Badezimmer:", energieverbrauch["Badezimmer"], "kWh")

# Gesamtenergieverbrauch berechnen
gesamtverbrauch = energieverbrauch["Wohnzimmer"] + energieverbrauch["Küche"] + energieverbrauch["Schlafzimmer"] + energieverbrauch["Badezimmer"]

# Gesamtergebnis anzeigen
print("Gesamtenergieverbrauch:", gesamtverbrauch, "kWh")
```

### b) Erweiterung mit Wiederholung
Wie könnte man dieses Programm vereinfache, so dass nicht jeder Eintrag in der Variable "energieverbrauch" einzeln ausgegeben werden muss? Versuche diese Erweiterung zu programmieren.

```python
# Energieverbrauchswerte der Räume in kWh
energieverbrauch = {
    "Wohnzimmer": 12.5,
    "Küche": 7.8,
    "Schlafzimmer": 5.2,
    "Badezimmer": 3.6
}

# Energieverbrauch für jeden Raum anzeigen
for raum, verbrauch in energieverbrauch.items():
    print(raum, ':', verbrauch, 'kWh')

# Gesamtenergieverbrauch berechnen
gesamtverbrauch = sum(energieverbrauch.values())

# Gesamtergebnis anzeigen
print("Gesamtenergieverbrauch:", gesamtverbrauch, "kWh")
```

## Aufgabe 5
Wir haben uns Gedanken dazu gemacht, wie wir Primzahlen in einem bestimmten Intervall identifizieren können. Bevor wir aber die beiden Algorithmen (Probedivisionen und Sieb des Eratosthenes) selber implementieren können, verwenden wir einen vorhandenen Algorithmus als Funktion, welche feststellt, ob eine Zahl prim ist oder nicht. Diese Funktion kannst du aus dem Modul [sympy](https://www.sympy.org/en/index.html) als Funktion _isprime_ importieren und verwenden.


```python
# in der Bibliothek "sympy" gibt es bereits einen Befehl "isprime", mit welchem wir die Frage beantworten können:
from sympy import isprime
zahl = int(input("Gib eine natürliche Zahl ein: "))

if isprime(zahl):
    print(zahl, " ist eine Primzahl.")
else:
    print(zahl, " ist keine Primzahl.")
```

## Aufgabe 6
Wir wollen mit Python einige elementare Mengenoperationen durchführen. Dazu sind die folgenden Mengen gegeben:
+   U = {1, 3, 5, 7, 9, 11, 13}
+   G = {2, 4, 6, 8, 10, 12, 14}
+   P = {1, 3, 5, 7, 11, 13, 17, 19, 23}

Bilde mit deinem erstellten Python-Programm die Vereinigungsmenge ($U \cup G$), Schnittmenge ($U \cap P$) und Differenzmenge ($P \setminus U$). Gib jeweils die Resultate auf dem Bildschirm aus.


```python
U = {1, 3, 5, 7, 9, 11, 13}
G = {2, 4, 6, 8, 10, 12, 14}
P = {1, 3, 5, 7, 11, 13, 17, 19, 23}

# Operationen auf Mengen
V = U.union(G)              # Vereinigungsmenge
S = U.intersection(P)       # Schnittmenge
D = P.difference(U)         # Differenzmenge

print("U =", U)
print("G =", G)
print("P =", P)

print("Vereinigungsmenge \nU \u222A G =", V)        # Operator für die Vereinigungsmenge \u222A (Unicode)
print("Schnittmenge \nU \u2229 P =", S)             # Operator für die Schnittmenge \u2229 (Unicode)
print("Differenzmenge \nP \u2216 U =", D)           # Operator für die Differenzmenge \u2216 (Unicode)
```

## Aufgabe 8
Wir wollen in Abhängigkeit der unterschiedlichen
Garstufen und der aktuellen Temperatur unseres Steak's, herausfinden, in welche
Stufe es eingeteilt werden kann und wie wir diese bezeichnen.
Die entsprechenden Garstufen sowie deren Bezeichnung findet man untenstehend:

| Garstufe-Deutsch | Garstufe-Englisch | Temperatur (Grad Celsius) |
|:-----------------|:------------------|:------------------------- |
| roh              | raw               |       T < 45              |
| blutig           | rare              | 45 <= T < 53              |
| rosa             | medium rare       | 53 <= T < 57              |
| halb durch       | medium            | 57 <= T < 63              |
| durch            | well done         | 63 <= T                   |

Erstelle ein Programm, welches dich nach der Temperatur abfragt und anschliessend die Garstufe ausgibt.

```python
# Temperatur abfragen
steak_temperatur = int(input('Gib die Zieltemperatur ein:'))

if steak_temperatur < 45:
    garstufe = "ROH / RAW"
elif steak_temperatur < 53:
    garstufe = "BLUTIG / RARE"
elif steak_temperatur < 57:
    garstufe = "ROSA / MEDIUM RARE"
elif steak_temperatur < 63:
    garstufe = "HALB DURCH / MEDIUM"
else:
    garstufe = "DURCH / WELL DONE"

print("Steaktemperatur:", steak_temperatur)
print("Garstufe:", garstufe)
```

## Aufgabe 9
Erstelle ein Pythonprogramm, welches dich nach den Koeffizienten (Fliesskommazahlen) einer quadratischen Gleichung fragt und anschliessend die Lösungen ausgibt, oder die Mitteilung, dass die Gleichung keine reellen Lösungen besitzt. 


```python
# für die Berechnung der Quadratwurzel benötigen wir das Modul "math"
import math as math

a = float(input("Gib die den Koeffizienten a ein:"))
b = float(input("Gib die den Koeffizienten b ein:"))
c = float(input("Gib die den Koeffizienten c ein:"))

diskriminante = b ** 2 - 4 * a * c

if diskriminante >= 0:
    if diskriminante == 0:
        loesung = - b / (2 * a)
        print("Die Doppellösung lautet:", loesung)
    else:
        loesung_1 = (- b + math.sqrt(diskriminante)) / (2 * a)
        loesung_2 = (- b - math.sqrt(diskriminante)) / (2 * a)
        print("Die beiden Lösungen lauten:")
        print("x1=", loesung_1)
        print("x2=", loesung_2)
else:
    print("Die Gleichung besitzt keine reellen Lösungen!")
    
```
