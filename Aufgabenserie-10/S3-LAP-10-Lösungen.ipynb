Bei einem Pkw wurden die CO-Leerlaufemission $x_i$ und der Kraftstoffverbrauch $y_i$ gemessen. Untersuche, ob ein linearer Zusammenhang besteht, d.h. ob die Vergasereinstellung den Kraftstoffverbrauch beeinflusst.

#### a) Gib die beiden Koeffizienten der Regressionsgerade aus

```python
import scipy.stats as stats
import numpy as np

X = np.array([2, 4, 5.6, 3.2, 2.8, 6, 3.5, 7.2, 8, 10])
Y = np.array([360, 380, 420, 380, 375, 392, 390, 416, 418, 424])

# Berechnung aller ben√∂tigten Paramter mit einem Befehl. Die Ausgabe
# der Regressionsgerade soll hier die positive Korrelation visualisieren.
m, a, r, p, e = stats.linregress(X,Y)

print("Steigung:", "%.3f " % m)
print("Achsenabschnitt:", "%.3f " % a)
#print("Korrelationskoeffizient:", "%.3f " % r)
```

### b) Stelle die Daten als Scatterplot dar

```python
import scipy.stats as stats
import numpy as np
import matplotlib.pyplot as plt

X = np.array([2, 4, 5.6, 3.2, 2.8, 6, 3.5, 7.2, 8, 10])
Y = np.array([360, 380, 420, 380, 375, 392, 390, 416, 418, 424])

# Berechnung aller ben√∂tigten Paramter mit einem Befehl. Die Ausgabe
# der Regressionsgerade soll hier die positive Korrelation visualisieren.
m, a, r, p, e = stats.linregress(X,Y)

plt.plot(X, Y,'bo')
plt.plot(X, m * X + a, color='red')
plt.xlabel("CO-Werte")
plt.ylabel("Kraftstoffverbrauch")
plt.title("Positive Korrelation - Regressionsgerade (rot)")

plt.show()
```

### c) Berechne einen Sch√§tzwert (Prognose)
1.  _Schritt_: Berechnung der Regressionsparameter
    -   m: Steigung der Regressionsgeraden
    -   a: Achsenabschnitt der Regressionsgeraden.
  
2.  _Schritt_: Prognose erstellen mit Hilfe der berechneten, linearen Funktion (Regressionsgerade). 
    -   Der Wert von $y$ wird durch Einsetzen von $ùë• = 12$ in die Gleichung $ùë¶ = ùëö‚ãÖùë• + ùëé$ berechnet.

3.  _Schritt_: Die prognostizierte Ausgabe wird auf zwei Dezimalstellen formatiert.

```python
import scipy.stats as sci
import numpy as np

# Eingabedaten
X = np.array([2, 4, 5.6, 3.2, 2.8, 6, 3.5, 7.2, 8, 10])
Y = np.array([360, 380, 420, 380, 375, 392, 390, 416, 418, 424])

# Berechnung der Regressionsparameter
m, a, r, p, e = sci.linregress(X, Y)

# Prognose f√ºr x = 12
x_prognose = 12
y_prognose = m * x_prognose + a

print(f"Prognostizierter Wert f√ºr x={x_prognose}: y={y_prognose:.2f}")
```

## Aufgabe 2 - Lineare Regression, Daten einlesen

Schreibe ein Python-Skript, das Messdaten zu Spannung (U) und Strom (I) aus einer Datei einliest und durch eine lineare Regression den Zusammenhang visualisiert.

-   Zeige das Diagramm an und gib die Regressionskoeffizienten in der Konsole aus. Ziel ist es, die Beziehung zwischen Spannung und Strom zu analysieren und visuell darzustellen.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Einlesen der Daten mit den Parametern: 
# 1) sep:       Trenzzeichen zwischen Spalten
# 2) decimal:   Dezimalzeichen
# 3) index_col: soll ein Index beim Importieren erstellt werden oder nicht
daten = pd.read_csv('Daten/daten-ein-10-2.txt', sep = ';', decimal = '.', index_col = False)

# Spannungsdaten und Stromdaten in Variablen X und Y √ºbernehmen
X = daten['U']
Y = daten['I']

# Visualisierung erstellen
plt.scatter(X, Y)

# Trendlinie, Grad des Polynoms ist 1, also eine Gerade
koeffizienten = np.polyfit(X, Y, 1)     # die Koeffizienten (a, b) ermitteln
f = np.poly1d(koeffizienten)            # ein Polynom f(x)= ax + b erstellen

print('Koeffizienten =', koeffizienten)
print('Polynom f(x)=', f)

# Trendlinie zum Plot hinzuf√ºgen
plt.plot(X, f(X), color = "red", linewidth = 1, linestyle = '-')

# Beschriftungen
plt.title('Scatterplot und Regressionsgerade (Trendlinie)')
plt.xlabel('U (Spannung in V)')
plt.ylabel('I (Stromwerte in A)')

# Grafik anzeigen
plt.show()
```

## Aufgabe 5 - Lineare Regression (polyfit)
#### a) Lineares Modell
Es werden 25 normalverteilte Zufallszahlen als x-Werte bestimmt. Die Lage und die Streuung werden als Paramterwerte vorgegeben. Die y-Werte zu diesen x-Werten sind durch eine quadratische Funktion vorgegeben. Dies entspricht unserer Realit√§t. Wir wollen nun durch eine __lineare Regression__ ein Modell berechnen, welche die angegebenen Messpunkte m√∂glichst gut ann√§hert.

```python
import scipy.stats as stats
import numpy as np
from sklearn.metrics import r2_score
import matplotlib.pyplot as plt

# Mittelwert
lage     = 5           # Mittelwert
standabw = 2           # Standardabweichung
anzahl   = 25          # Anzahl Zufallszahlen

# Daten zuf√§llig generieren mit den oben stehenden Parametern
x = np.random.normal(lage, standabw, anzahl)
y = -3 * x**2 + 4 * x + 1

# Bestimme die lineare Regression 
# (Polynom mit Grad 1)
m, a = np.polyfit(x, y, 1)
# jetzt noch die X-Werte in die Regressionsgerade einsetzen:
y_fit_lin = m * x + a

# Bestimmtheitsmass
r_squared_lin = r2_score(y, y_fit_lin)

print("Lineares Modell: Bestimmtheitsmass: ", "%.4f " % r_squared_lin)
print("Lineares Modell: Koeffizienten: Steigung:", "%.4f " % m, "Achsenabschnitt:", "%.4f " % a)
```

#### b) Quadratisches Modell
Es werden 25 normalverteilte Zufallszahlen als x-Werte bestimmt. Die Lage und die Streuung werden als Paramterwerte vorgegeben. Die y-Werte zu diesen x-Werten sind durch eine quadratische Funktion vorgegeben. Dies entspricht unserer Realit√§t. Wir wollen nun durch eine __polynomiale Regression__ ein Modell berechnen, welche die angegebenen Messpunkte m√∂glichst gut ann√§hert.

```python
import numpy as np
from sklearn.metrics import r2_score
import matplotlib.pyplot as plt

# Mittelwert
lage     = 5           # Mittelwert
standabw = 2           # Standardabweichung
anzahl   = 25          # Anzahl Zufallszahlen

# Daten zuf√§llig generieren mit den oben stehenden Parametern
x = np.random.normal(lage, standabw, anzahl)
y = -3 * x**2 + 4 * x + 1

# Bestimme die Anpassung mit einer quadratischen Funktion 
# (Polynom mit Grad 2)
koeff = np.polyfit(x, y, 2)
a = koeff[0]
b = koeff[1]
c = koeff[2]

# jetzt noch die x-Werte in die Kurve einsetzen:
y_fit_quad = a * x**2 + b * x + c

# Bestimmtheitsmass
r_squared_quad = r2_score(y, y_fit_quad)
print("Quadratisches Modell: Bestimmtheitsmass: ", "%.4f " % r_squared_quad)
print("Quadratisches Modell: Koeffizienten:", "a=", "%.4f " % a, "b=", "%.4f " % b, "c=", "%.4f " % c)
```

Wie wir beim Vergleich feststellen, ist das Bestimmtheismass beim polynomialen Modell nahezu perfekt, da der Wert 1 betr√§gt. Die berechneten Modellkoeffizienten entsprechen genau denjenigen, welche wir beim Generieren der y-Werte benutzt haben. Durch diese einfache Benutzung der Methode `numpy.polyfit(x,y,Grad)` erhalten wir ein m√§chtiges Werkzeug um Messdaten durch ein beliebiges Polynom anzun√§hern oder m√∂glichst realit√§tsgetreu abzubilden.
