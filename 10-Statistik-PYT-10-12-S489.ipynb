```python
# Befehle aus der Bibliothek numpy importieren
import numpy as np

# Statistik-Befehle aus scipy importieren
from scipy import stats

# Befehle zum Plotten von Datenpunkten importieren
import matplotlib.pyplot as plt
```

```python
x = np.array([2, 4, 5.6, 3.2, 2.8, 6, 3.5, 7.2, 8, 10])
```

```python
y = np.array([360, 380, 420, 380, 375, 392, 390, 416, 418, 424])
```

Wir berechnen die benoetigten Paramter unserer Regressionsgeraden mit einem Befehl. Es duerfen keine Parameter weggelassen werden, obschon nicht alle fuer die Loesung benoetigt werden. Der Befehl stats.linregress(x,y) liefert die Parameter: (m) Steigung, (a) Achsenabschnitt, (r) Korrelationskoeffizient, (p) P-Wert, (e) Standardfehler. Eigentlich gibt es sechs Parameter. Wegen Kompatibilitaetsgruenden werden nur fünf verlangt. Der sechste Parameter kann aber manuell ermittelt werden:
https://docs.scipy.org/doc/scipy-1.7.1/reference/generated/scipy.stats.linregress.html

```python
m, a, r, p, e = stats.linregress(x,y)
```

Der Befehl stats.linregress(x,y) liefert die Parameter:
m: Steigung
a: Achsenabschnitt
r: Korrelationskoeffizient
p: P-Wert
e: Standardfehler

```python
plt.title ('Beispiel 10.14, S.491')
plt.plot(x, y, 'rx', label='Messwerte')
plt.plot(x, m*x + a, label='Schaetzung: Regressionsgerade')
plt.xlabel("CO-Leerlaufemissionen")
plt.ylabel("Kraftstoffverbrauch")
plt.legend()
plt.show()
```

Der lineare Zusammenhang wird durch den Korrelationskoeffizienten (r-value) ausgedrueckt. Dieser liegt im Intervall [-1;1]. Falls er den Wert 0 annimmt, dann gibt es keinen linearen Zusammenhang.
