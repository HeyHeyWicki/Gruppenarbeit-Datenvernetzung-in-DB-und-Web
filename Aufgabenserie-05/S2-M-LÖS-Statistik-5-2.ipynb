## Aufgabe 2 - Daten einer Zeitreihe erstellen

Zusammenfassung beim Erstellen der Daten:
1. Trend: Eine langsame, gleichmäßige lineare Zunahme der Werte.
2. Saisonale Komponente: Eine periodische Schwankung innerhalb von einem Jahr (365 Tagen), z. B. durch jahreszeitliche Effekte.
3. Zyklische Komponente: Langsamere, mehrjährige Schwankungen, die größere Zyklen beschreiben.
4. Restkomponente: Zufällige, unvorhersehbare Schwankungen.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Tage (1 Jahr = 365 Tage)
t = np.arange(1, 366)

# Trend-Komponente (T)
T = 0.05 * t

# Saisonale Komponente (S) - 1-jähriger Zyklus
S = 10 * np.sin(2 * np.pi * t / 365)

# Zyklische Komponente (Z) - 2-jähriger Zyklus
Z = 5 * np.sin(2 * np.pi * t / 730)

# Rauschkomponente (R) - zufälliges Rauschen
np.random.seed(42)  # Für reproduzierbare Ergebnisse
R = np.random.normal(0, 2, len(t))

# Gesamte Zeitreihe (Y)
Y = T + S + Z + R

# Erstellen eines Pandas-Datenrahmens
df = pd.DataFrame({
    'Tag': t,
    'Trend': T,
    'Saison': S,
    'Zyklus': Z,
    'Rest': R,
    'Gesamt': Y
})

# Erste 10 Zeilen des Datensatzes anzeigen
print(df.head(10))

# Visualisierung der Zeitreihe
plt.figure(figsize=(10, 6))
plt.plot(t, Y, label="Gesamt")
plt.plot(t, T, label="Trend", linestyle='--')
plt.plot(t, S, label="Saison", linestyle='--')
plt.plot(t, Z, label="Zyklus", linestyle='--')
plt.legend()
plt.title("Zeitreihe mit Trend, Saison, Zyklus und Rest")
plt.xlabel("Tag")
plt.ylabel("Wert")
plt.show()
```
