## Aufgabe 1 - Lichtsteuerung
Implementiere ein Programm, das die Beleuchtung basierend auf der Uhrzeit steuert. Zwischen 18:00 und 06:00 Uhr soll das Licht eingeschaltet sein, sonst ausgeschaltet. Der Benutzer soll eine Eingabe der Uhrzeit (ganze Stunden) vornehmen können und du sollst dafür eine Funktion _lichtsteuerung_ implementieren.

```python
def lichtsteuerung(zeit):
    start_zeit_1 = 18       # 18:00 Uhr, abends
    end_zeit_1   = 24       # 24:00 Uhr, Mitternacht
    start_zeit_2 = 0        # 00:00 Uhr, Mitternacht
    end_zeit_2   = 6        # 06:00 Uhr, morgens      
    
    if (start_zeit_1 <= zeit <= end_zeit_1) or (start_zeit_2 <= zeit < end_zeit_2):
        return "Licht an"
    else:
        return "Licht aus"

uhrzeit = int(input("Gib die aktuelle Uhrzeit (0-24) ein: "))

print(lichtsteuerung(uhrzeit))
```

## Aufgabe 2
### a) Thermostat mit Zieltemperatur
Implementiere eine Funktion _thermostat_, die die aktuelle Temperatur mit einer Zieltemperatur vergleicht und entsprechend reagiert. das heisst, die Nachricht "Heizung an" soll ausgegeben werden wenn, die aktuelle Temperatur kleiner als die Zieltemperatur ist. Umgekehrt, soll der Text "Kühlung an" ausgegeben werden. Falls die beiden Temperaturwerte übereinstimmen, soll die Meldung "Keine Aktion erforderlich" auf dem Bildschirm ausgegeben werden. Die Zieltemperatur soll ebenfalls zur aktuellen Temperatur eingegeben werden können.


```python
def thermostat(aktuelle_temp, ziel_temp):
    if aktuelle_temp < ziel_temp:
        return "Heizung an"
    elif aktuelle_temp > ziel_temp:
        return "Kühlung an"
    else:
        return "Keine Aktion erforderlich"

akt_temp = float(input("Gib die aktuelle Temperatur ein: "))
ziel_temp = float(input("Gib die Zieltemperatur ein: "))

print(thermostat(akt_temp, ziel_temp))
```

### b) Thermostat mit Zieltemperatur, überwacht
Erweitere dein Programm so, dass wiederholt die Temperaturkontrolle (Funktion: _temperatur_kontrolle_) durchgeführt wird, bis eine akzeptable Temperatur (zwischen 20°C und 25°C) erreicht ist. Gib bei jeder Eingabe die aktuelle Raumtemperatur und den Status aus.

```python
def temperatur_kontrolle(temperatur):
    min_temp = 20
    max_temp = 25
    if temperatur < min_temp:
        return "Heizung an"
    elif temperatur > max_temp:
        return "Kühlung an"
    else:
        return "Alles in Ordnung"

while True:
    aktuelle_temperatur = float(input("Gib die aktuelle Temperatur ein: "))
    status = temperatur_kontrolle(aktuelle_temperatur)
    print("Temperatur:", aktuelle_temperatur, "Status:", status)
    if status == "Alles in Ordnung":
        break   # falls das gewünschte Temperaturintervall erreicht wird, wird die Schleife verlassen
```

## Aufgabe 3 - Rollladensteuerung
Programmiere eine Funktion _roll_laden_steuerung_, die Rollladen basierend auf der Helligkeit (in Lux) steuert. Solange die Helligkeit in Lux kleiner ist als 300, soll die Meldung ercheinen: "Rollladen öffnen". Falls der Helligkeitswert grösser als 700 Lux beträgt, soll die Meldung "Rollladen schliessen" und sonst "Rolladen bleibt unverändert" ausgegeben werden. Der Benutzer soll die Helligkeit (natürliche Zahl) in Lux eingeben können.


```python
def roll_laden_steuerung(helligkeit):
    oeffnen = 300
    schliessen = 700
    if helligkeit < oeffnen:
        return "Rolladen öffnen"
    elif helligkeit > schliessen:
        return "Rolladen schliessen"
    else:
        return "Rolladen bleibt unverändert"

lux = int(input("Gib den Helligkeitswert (in Lux) ein: "))
print(roll_laden_steuerung(lux))
```

## Aufgabe 4 - Wasseralarm
Erstelle ein Pythonpogramm, das basierend auf Sensordaten (Wassererkennung als `True` oder `False`) einen Alarm ausgibt. Erstelle eine Funktion _wasser_alarm_, welche einen Eingabeparamter für die Wassererkennung aufweist. Der Benutzer soll eingeben können, ob Wassereinbruch erkannt worden ist oder nicht.

```python
def wasser_alarm(wassererkennung):
    if wassererkennung:
        return "Wasseralarm ausgelöst"
    else:
        return "Kein Wassereinbruch"

sensor = input("Ist Wasser erkannt worden? (ja/nein): ").lower() == "ja"
print(wasser_alarm(sensor))
```

## Aufgabe 5 - Wasseralarm mit Zufallszahlen simuliert
Erweitere dein erstelltest Programm aus der vorherigen Aufgabe so, dass durch Generierung von Zufallszahlen, der Wasseralarm ausgelöst wird oder nicht.

```python
import random
import time

def wasser_alarm(wassererkennung):
    if wassererkennung:
        return "Wasseralarm ausgelöst"
    else:
        return "Kein Wassereinbruch"

# Schleife, um die Wassererkennung regelmässig zu überprüfen
while True:
    # zufälligen booleschen Werte mit Hilfe von random.choice bestimmen
    wasser_erkannt = random.choice([True, False])
    print(wasser_alarm(wasser_erkannt))
    if wasser_erkannt:
        break  # Beendet die Schleife, wenn Wasser erkannt wurde
    time.sleep(2)  # Simuliert eine Pause von 2 Sekunden zwischen den Überprüfungen
```

## Aufgabe 6 - Lüftung
Das Programm soll kontinuierlich die Luftqualität überwachen und die Lüftung so lange einschalten, bis ein sicherer Wert erreicht ist (kleiner oder gleich 1000 ppm). Verwende dazu eine Funktion _lueftung_ueberwachung_. Die zu verwendenden Stati lauten: "Lüftung an", "Lüftung kann ausgeschaltet werden".

```python
def lueftung_ueberwachung(ppm_wert):
    untere_grenze = 1000
    if ppm_wert > untere_grenze:
        return "Lüftung an"
    else:
        return "Lüftung kann ausgeschaltet werden"

while True:
    ppm = int(input("Geben Sie den aktuellen ppm-Wert der Luft ein: "))
    status = lueftung_ueberwachung(ppm)
    print(status)
    if status == "Lüftung kann ausgeschaltet werden":
        break
```

## Aufgabe 7 - Bewegungsmelder
Das System (Pythonprogramm) soll kontinuierlich überprüfen, ob eine Bewegung erkannt wird. Wenn fünf aufeinanderfolgende Abfragen (in der Realität fünf Minuten) ohne Bewegung stattfinden, wird das Licht ausgeschaltet. Erstelle dazu ein Pythonprogramm mit einer Funktion _licht_bei_bewegung_, welche einen Eingabeparameter verwendet. Dieser gibt an, ob eine Bewegung erkannt worden ist oder nicht.

```python
def licht_bei_bewegung(bewegung):
    if bewegung:
        return "Licht an"
    else:
        return "Licht aus nach 5 Minuten"

keine_bewegung_zaehler = 0
licht_aus_zaehler      = 5

while keine_bewegung_zaehler < licht_aus_zaehler:
    bewegung_erkannt = input("Bewegung erkannt? (ja/nein):")
    if bewegung_erkannt == "ja":
        # Funktionsaufruf mit erkannter Bewegung
        print(licht_bei_bewegung(True))
        keine_bewegung_counter = 0  # Zähler muss zurückgesetzt werden!
    else:
        keine_bewegung_counter += 1
        print("Keine Bewegung erkannt. Zähler:", keine_bewegung_counter)
        if keine_bewegung_counter >= licht_aus_zaehler:
            # Funktionsaufruf, keine Bewegung erkannt
            print(licht_bei_bewegung(False))
            # verlässt die Wiederholungsschleife sofort,
            # das Licht wird nach fünf Minuten ausgeschaltet.
            break
```

## Aufgabe 8 - Filterwechsel
Das Programm soll alle 100 Betriebsstunden, beginnend bei 0 die verbleibenden Stunden bis zum Filterwechsel überprüfen, ob der Filterwechsel erforderlich ist. Durch Verwendung einer Schleife soll in jedem Durchgang die Betriebsstundenanzahl um 100 erhöht werden. Falls die Anzahl der Betriebsstunden grösser als 1000 Stunden ist, soll der Filter ausgewechselt werden. Verwende dazu eine Funktion _filterwechsel_benachrichtigung_, welche die Anzahl Betriebsstunden als Parameter verwendet.

```python
def filterwechsel_benachrichtigung(betriebsstunden):
    if betriebsstunden >= filterwechsel_stunden:
        return "Filterwechsel erforderlich"
    else:
        return f"Noch {filterwechsel_stunden - betriebsstunden} Betriebsstunden bis zum Filterwechsel"

filterwechsel_stunden = 1000    # globale Variable
aktuelle_stunden = 0

while aktuelle_stunden < filterwechsel_stunden:
    print(filterwechsel_benachrichtigung(aktuelle_stunden))
    aktuelle_stunden += 100  # Simuliert weitere 100 Stunden Betrieb
    input("Drücke Enter, um die nächste Überprüfung zu simulieren.")

print("Filterwechsel erforderlich")
```

## Aufgabe 9 - Rohlingsfläche
In einem Betrieb sollen Rohlinge (Kreiszylinder; hohl) aus Blech ausgeschnitten werden. Damit die Maschine weiss, wieviel Fläche sie aus dem Blech ausschneiden muss, wird ein Pythonprogramm erstellt. Dieses soll vom Benutzer den Radius R und die Höhe H des Kreiszylinders abfragen. Anschliessend soll die Mantelfläche mit Hilfe einer Funktion _mantel_flaeche_ berechnet und ausgegeben werden. Runde das Resultat auf drei Nachkommastellen.


```python
import math as math     # für die Konstante Pi

def mantel_flaeche (radius, hoehe):
    umfang = 2 * radius * math.pi
    mantel = umfang * hoehe
    return mantel

r = float(input("Gib den Radius des Kreiszylinders ein:"))
h = float(input("Gib die Höhe des Kreiszylinders ein:"))

print("Die Mantelfläche des Kreiszylinders beträgt:", round(mantel_flaeche(r, h),3))
```
